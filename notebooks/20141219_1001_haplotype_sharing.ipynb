```python
%%px --local
import os, json
import pandas as pd
import numpy as np
import scipy
import pickle
#from numpy import *
import matplotlib as mpl
from matplotlib import pyplot as plt
eu = os.path.expanduser
jn = os.path.join
mpl.rcParams["font.size"]=14
eu = os.path.expanduser
jn = os.path.join
meta_dir = eu("~/1001genomes_project/Hannes/metadata")
data_dir = eu("~/1001genomes_project/intermediate/fullGenomes3")
mt = pd.read_csv(eu("~/1001genomes_project/Hannes/metadata/1135_master_table_admix_pop.csv"),index_col=0)
seq_res = pd.read_csv(eu("~/1001genomes_project/Hannes/metadata/Overview_of_sequencing_results.csv"),index_col=0,skiprows=2)
```

```python
def get_ibd_df(chrom):
    ibd_df = pd.read_csv(jn(data_dir,"1001g_fG3_1135_biallelic_snp_beagle_Chr{}.ibd".format(chrom)),
                     sep="\t",names=["id0","strd0","id1","strd1","chrom","start","end","lod"])
    ibd_df = ibd_df.drop_duplicates()
    ibd_df["length"] = ibd_df["end"] -ibd_df["start"]
    ibd_df["pop0"] = [s for s in  mt.ix[ibd_df["id0"]]["population"].values]
    ibd_df["pop1"] = [s for s in  mt.ix[ibd_df["id1"]]["population"].values]
    ibd_df["c0"] = [s for s in  mt.ix[ibd_df["id0"]]["country"].values]
    ibd_df["c1"] = [s for s in  mt.ix[ibd_df["id1"]]["country"].values]
    return ibd_df
```

```python
ibd_map = lv.map_async(get_ibd_df,chromosomes)
```

```python
ibd_df = pd.concat(ibd_map.result,ignore_index=True)
```

```python
pop_group = ibd_df.groupby(["pop0","pop1"])
```

```python
l = pop_group.apply(lambda x: len(x)).unstack().fillna(0)
l = (l+l.T)
l_df = pd.DataFrame(np.triu(l.values),index=l.index,columns=l.columns)
vc = mt["population"].value_counts()
#this is the number of individual pairs, makes a small error for diagonal entries
normalisation = pd.DataFrame(np.outer(vc.values,vc.values)/2,index=vc.index,columns=vc.index)
print "number of shared ibd_tracks"
(l_df/normalisation).applymap(round)
```

```python
max_len = pop_group.apply(lambda df: df["length"].max()).unstack()
print "maximum length of shared ibd tracks in kb"
(max_len/1000).applymap(round)
```

```python
med_len = pop_group.apply(lambda df: df["length"].median()).unstack()
print "median length of shared ibd tracks in kb"
(med_len/1000).applymap(round)
```
