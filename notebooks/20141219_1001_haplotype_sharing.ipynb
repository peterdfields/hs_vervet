```python
def get_ibd_df(chrom):
    ibd_df = pd.read_csv(jn(data_dir,"1001g_fG3_1135_biallelic_snp_beagle_Chr{}.ibd".format(chrom)),
                     sep="\t",names=["id0","strd0","id1","strd1","chrom","start","end","lod"])
    ibd_df = ibd_df.drop_duplicates()
    ibd_df["length"] = ibd_df["end"] -ibd_df["start"]
    ibd_df["pop0"] = [s for s in  mt.ix[ibd_df["id0"]]["population"].values]
    ibd_df["pop1"] = [s for s in  mt.ix[ibd_df["id1"]]["population"].values]
    ibd_df["c0"] = [s for s in  mt.ix[ibd_df["id0"]]["country"].values]
    ibd_df["c1"] = [s for s in  mt.ix[ibd_df["id1"]]["country"].values]
    return ibd_df
```

```python
def get_ibd_matrices(chrom): 
    ibd_df = pd.read_csv(jn(data_dir,"1001g_fG3_1135_biallelic_snp_beagle_Chr{}.ibd".format(chrom)),
                         sep="\t",names=["id0","strd0","id1","strd1","chrom","start","end","lod"])
    ibd_df["length"] = ibd_df["end"] -ibd_df["start"]
    ibd_df = ibd_df.drop_duplicates()
    ig = ibd_df.groupby(['id0','id1'])
    tot_length_mat = ig.apply(lambda df: df['length'].sum()).unstack().fillna(0)
    tot_length_mat.index = pd.MultiIndex.from_tuples(zip(mt.ix[tot_length_mat.index]["population"],
                                                            mt.ix[tot_length_mat.index]["country"],
                                                            tot_length_mat.index))
    tot_length_mat.columns = pd.MultiIndex.from_tuples(zip(mt.ix[tot_length_mat.columns]["population"],
                                                            mt.ix[tot_length_mat.columns]["country"],
                                                            tot_length_mat.columns))
    tot_length_mat.sort_index(inplace=True)
    count_mat = ig.apply(lambda df: len(df)).unstack().fillna(0)
    count_mat.index = pd.MultiIndex.from_tuples(zip(mt.ix[count_mat.index]["population"],
                                                            mt.ix[count_mat.index]["country"],
                                                            count_mat.index))
    count_mat.columns = pd.MultiIndex.from_tuples(zip(mt.ix[count_mat.columns]["population"],
                                                            mt.ix[count_mat.columns]["country"],
                                                            count_mat.columns))
    count_mat.sort_index(inplace=True)
    return tot_length_mat, count_mat
```

```python
map_pairwise_comp = lv.map_async(get_ibd_matrices,chromosomes)
```

```python
def reduce_pairwise_comp(result):
    length_mat = sum([c[0].fillna(0) for c in result])
    count_mat = sum([c[1].fillna(0) for c in result])
    return length_mat, count_mat
```

```python
def reduce_pairwise_comp(result):
    length_mat = result[0][0]
    count_mat = result[0][1]
    for r in result[1:]:
        length_mat.add(r[0],fill_value=0)
        count_mat.add(r[1],fill_value=0)
    return length_mat, count_mat
```
