```python
import os, json
import pandas as pd
from matplotlib import pyplot as plt
eu = os.path.expanduser
jn = os.path.join
meta_dir = eu("~/vervet_project/metadata")
align_siv_ana_dir = eu("~/vervet_project/analyses/20141017_align_mito_siv/_data")
```

```python
import ast
def str_to_dic(s):
    try:
        return ast.literal_eval(s)
    except ValueError:
        return {}
mt = pd.read_csv(jn(meta_dir,"163_master_table.csv"),
                 index_col=0,
                 converters={"library":str_to_dic,"phenotypes":str_to_dic}
                 )
```

```python
cov_df = pd.read_table(jn(align_siv_ana_dir,"coverage_mito_siv"),index_col=[0,1],header=None,names=mt.index)
```

```python
#normalised by average autosomal coverage
mean_cov = cov_df.groupby(lambda idx:idx[0]).mean().div(mt["coverage"],axis=1).T
del mean_cov["gi|107736076|ref|NC_008066.1|"]
```

```python
mean_cov.ix[">DJ048201|DJ048201"].hist(bins=50)
```

```python
#mapping against all viral sequences is perfactly correlated
plt.scatter(mean_cov.ix[">DJ048201|DJ048201"],mean_cov.ix[">HQ378594|HQ378594"])
```

```python
# work with mean across all viral sequences
mean_cov = mean_cov.mean(axis=1)
```

```python
mt.country.value_counts()
```

```python
country_mean = mean_cov.groupby(mt.country).mean()
country_std = mean_cov.groupby(mt.country).std()
```

```python
ax = plt.gca()
ax.errorbar(range(len(country_mean)), country_mean, yerr=country_std, fmt='o')
plt.xticks(range(len(country_mean)),country_mean.index, rotation='vertical')
ax.set_xlim([-1,len(country_mean)])
```

```python
countries = mt.country.unique()
```

```python
mean_cov["a"] = 1
```

```python
country_num = pd.Series(index=mean_cov.index) 
for i,country in enumerate(countries):
    country_num.ix[mt["country"]==country] = i
```

```python
ax = plt.gca()
#plt.plot(country_num,mean_cov,".")
plt.scatter(country_num,mean_cov)
plt.xticks(range(len(countries)),countries, rotation='vertical')
ax.set_xlim([-1,len(countries)])
ax.set_ylim([0,0.6])
plt.title("normalised SIV coverage")
```
