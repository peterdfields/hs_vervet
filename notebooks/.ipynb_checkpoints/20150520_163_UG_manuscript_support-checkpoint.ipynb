```python
%%px --local
#general functions
def get_gen_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)

def get_subpop_df(gen_df,pop,poptype='pop3'):
    gen_df_pop = gen_df[mt[(mt[poptype]==pop)].index]
    #print len(gen_df_pop)
    #removing non-seggretating sites
    gen_df_pop = gen_df_pop[(gen_df_pop.mean(axis=1)>0)&(gen_df_pop.mean(axis=1)<2)]
    #print len(gen_df_pop)
    return gen_df_pop

def get_id_order(chrom):
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    with open(fn) as f:
        ids = f.readline().strip().split('\t')[2:]
    return ids


def get_ad_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_anc_der_including_fixed_{}.tsv".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)

def get_ad_seg_df(chrom):
    df = get_ad_df(chrom)
    return df[(df.sum(axis=1)>0)&\
                        (df.sum(axis=1)<2*df.notnull().sum(axis=1))]
```

```python
from matplotlib import markers

def align_marker(marker, halign='center', valign='middle',):
    """
    create markers with specified alignment.

    Parameters
    ----------

    marker : a valid marker specification.
      See mpl.markers

    halign : string, float {'left', 'center', 'right'}
      Specifies the horizontal alignment of the marker. *float* values
      specify the alignment in units of the markersize/2 (0 is 'center',
      -1 is 'right', 1 is 'left').

    valign : string, float {'top', 'middle', 'bottom'}
      Specifies the vertical alignment of the marker. *float* values
      specify the alignment in units of the markersize/2 (0 is 'middle',
      -1 is 'top', 1 is 'bottom').

    Returns
    -------

    marker_array : numpy.ndarray
      A Nx2 array that specifies the marker path relative to the
      plot target point at (0, 0).

    Notes
    -----
    The mark_array can be passed directly to ax.plot and ax.scatter, e.g.::

        ax.plot(1, 1, marker=align_marker('>', 'left'))

    """

    if isinstance(halign, (str, unicode)):
        halign = {'right': -1.,
                  'middle': 0.,
                  'center': 0.,
                  'left': 1.,
                  }[halign]

    if isinstance(valign, (str, unicode)):
        valign = {'top': -1.,
                  'middle': 0.,
                  'center': 0.,
                  'bottom': 1.,
                  }[valign]

    # Define the base marker
    bm = markers.MarkerStyle(marker)

    # Get the marker path and apply the marker transform to get the
    # actual marker vertices (they should all be in a unit-square
    # centered at (0, 0))
    m_arr = bm.get_path().transformed(bm.get_transform()).vertices

    # Shift the marker vertices for the specified alignment.
    m_arr[:, 0] += halign / 2
    m_arr[:, 1] += valign / 2

    return m_arr
```

```python
def to_wot(ax):
    """
    white axis (on transparent)
    """
    ax.tick_params(axis='x', colors='w')
    ax.tick_params(axis='y', colors='w')
    ax.yaxis.label.set_color('w')
    ax.xaxis.label.set_color('w')
    title_obj = ax.get_title()
    if title_obj and title_obj is not None:
        plt.setp(title_obj, color='w')
    l = ax.get_legend() 
    if l is not None:
        for text in l.get_texts():
            text.set_color("w")
    #cbytick_obj = plt.getp(colorbar.ax.axes, 'yticklabels')                #tricky
    #plt.setp(cbytick_obj, color='w')
    for child in ax.get_children():
        if isinstance(child, mpl.spines.Spine):
            child.set_color("white")
```

```python
def plot_map(fig,coastlineargs=None,wot=False):
    """
    wot .. white on transparent
    """
    if coastlineargs is None:
        coastlineargs = {'linewidth':0.5}
    if wot:
        coastlineargs.update({'color':'white'})
    ax = plt.gca()
    m = Basemap(llcrnrlon=-22 ,llcrnrlat=-39, urcrnrlon= 53,urcrnrlat=28, resolution='c',ax=ax,)
    m.drawcoastlines(**coastlineargs)
    l0,b0,w0,h0 = ax.get_position().bounds
    l = l0 + 0.07 * w0
    b = b0 + 0.3* h0
    w = 0.25 * w0
    h = 0.25 * h0
    ax_kitts = plt.axes([l, b, w, h])
    m_kitts = Basemap(llcrnrlon= -62.935123,llcrnrlat=17, urcrnrlon=-62.5 ,urcrnrlat=17.46, resolution='c',ax=ax_kitts)
    m_kitts.drawcoastlines(**coastlineargs)
    ax_kitts.set_xlabel("St. Kitts and Nevis",fontsize=12)
    ax_kitts.xaxis.set_label_position('top') 

    l = l0 + 0.07 * w0
    b = b0 + 0.07* h0
    w = 0.18 * w0
    h = 0.18 * h0
    ax_barb = plt.axes([l, b, w, h])
    m_barb = Basemap(llcrnrlon= -59.698642,llcrnrlat=13.017424, urcrnrlon=-59.37 ,urcrnrlat=13.369400, resolution='c',ax=ax_barb)
    m_barb.drawcoastlines(**coastlineargs)
    ax_barb.set_xlabel("Barbados",fontsize=12,color='k')
    ax_barb.xaxis.set_label_position('top') 
    
    if wot:
        to_wot(ax)
        to_wot(ax_kitts)
        to_wot(ax_barb)
    return ax, ax_kitts, ax_barb
```

### Site frequency spectra

```python
%%px --local
#data functions
def get_ac(gen_df,mode):
    """
    mode = "sum" ... just sums up the entries, this means that nans are treated as 0
    mode = "mean" ... calculates average ignoring nans -> the spectrum is pulled towards
                        extreme frequencies
    mode = "absorbing" ... pull the spectrum to the left for frequencies <0.5 and to the
                            right for frequencies >0.5
    """
    n = gen_df.shape[1]
    if mode == "sum":
        return gen_df.sum(axis=1)
    elif mode == "mean":
        af = gen_df.mean(axis=1)/2.
        return (af*2*n).round()
    elif mode == "absorbing":
        mean_gt0p5 = (gen_df.mean(axis=1)/2.>0.5)
        na_val = 2.*mean_gt0p5
        gen_df[gen_df.isnull().mul(mean_gt0p5,axis=0)]=2
        gen_df[gen_df.isnull().mul(~mean_gt0p5,axis=0)]=0
        return gen_df.sum(axis=1)
    else:
        raise

def get_acs(gen_df,mode):
    """
    get allele count spectrum (histogram)
    """
    n = gen_df.shape[1]
    ac = get_ac(gen_df,mode=mode)
    bins = np.arange(-0.5,2*n+0.501)
    h, bins = np.histogram(ac,bins=bins)
    return h, bins
    
def acs_from_chrom(chrom,mode="mean",anc_der=True):
    """
    allele count spectrum (histogram)
    for either ancestral-derived (anc_der=True)
    or ref-alt alleles
    """
    if anc_der:
        df = get_ad_df(chrom)
    else:
        df = get_gen_df(chrom)
    return get_acs(df,mode=mode)
```

```python
#plot function
def plot_hist(h,bins,legend=True,x0_waterson=2):
    
    width = 0.7 * (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    plt.bar(center, h, align='center', width=width,color="black")
    ax = plt.gca()
    ax.set_xlim(bins[0],bins[-1])
    
    #calculate and plot theta waterson
    n = int(bins[-1])
    theta_waterson = h.sum()*1./sum([1./i for i in range(1,n)])
    x = np.arange(x0_waterson,2*n)
    label ="neutral spectrum\n w/ wattersons theta"
    line, = plt.plot(x,theta_waterson/x,linewidth=2.5, linestyle="--",color=(0.8,0.8,0),
                label=label)
    if legend:
        ax.legend()
    return line,label
```

```python
%%px --local
#snps within an between, in numbers
def get_subspecies_af(gen_df):
    gen_df.columns = pd.MultiIndex.from_tuples(zip(mt.ix[gen_df.columns]["pop3"],
                                                            mt.ix[gen_df.columns]["country"],
                                                            gen_df.columns))
    ss_groups = gen_df.groupby(axis=1,level=0)
    #this is much faster than using the groupby mean function
    ss_df = pd.DataFrame()
    for g, df in ss_groups:
        ss_df[g] = df.mean(axis=1)/2  
    return ss_df

def get_unique_snps_per_group(ss_df,af_threshold):
    superset = [c for i in range(1,len(pops3)+1) for c in it.combinations(pops3,i)]
    snps_only_in = {}
    for focal, other in zip(superset,
                            ([pop for pop in pops3 if pop not in superset[i]] for i in range(len(superset)))): 
        snps_only_in.update({focal:((ss_df[list(focal)]>af_threshold).all(axis=1)&\
                                     (ss_df[other]<=af_threshold).all(axis=1)).sum()})
    shared_and_private_snps = pd.Series(snps_only_in,index=snps_only_in.keys())
    shared_and_private_snps.sort(ascending=False)
    return shared_and_private_snps

def get_snp_groups_chrom(chrom,af_threshold_ls):
    gen_df = get_gen_df(chrom)
    ss_df = get_subspecies_af(gen_df)
    return [get_unique_snps_per_group(ss_df,t) for t in af_threshold_ls]
```

```python
def reduce_snps_per_group(results):
    n_el = len(results[0])
    tot_res = [None for i in range(n_el)]
    for i in range(n_el):
        tot_res[i] = reduce(lambda x,y: x+y,[r[i] for r in results])
        tot_res[i].sort(ascending=False)
    return tot_res
```

### Subpopulation SFS

```python
%%px --local
def get_acs2D(gen_df1,gen_df2,mode):
    """
    get 2D allele count spectrum (histogram)
    for 2 populations
    """
    n1 = gen_df1.shape[1]
    n2 = gen_df2.shape[1]
    ac1 = get_ac(gen_df1,mode=mode)
    ac2 = get_ac(gen_df2,mode=mode)
    bins = [np.arange(-0.5,2*n1+0.501),np.arange(-0.5,2*n2+0.501)]
    h = np.histogram2d(ac1.values,ac2.values,bins=bins)
    return h

def get_pop_hists_1d2d(chrom,mode="mean",population='pop3'):
    """
    here Caribbean sabaeus are treated as 2 seperate populations
    """
    hists = {}
    hists2d = {}
    gen_df = get_ad_seg_df(chrom)
    if population == 'population':
        popnames = pops
    elif population == 'pop2':
        popnames = pops2
    elif population == 'pop3':
        popnames = pops3
    else:
        raise ValueError("population must be in ['population','pop2','pop3']")
    for i,pop in enumerate(popnames):
        p_gen_df = gen_df.ix[:,mt[mt[population]==pop].index.unique()]
        p_gen_df_non_fixed = p_gen_df[(p_gen_df.sum(axis=1)>0)&\
                                        (p_gen_df.sum(axis=1)<2*p_gen_df.notnull().sum(axis=1))]
        hists.update({pop:get_acs(p_gen_df_non_fixed,mode=mode)})
        for pop2 in popnames[i+1:]:
            p2_gen_df = gen_df.ix[:,mt[mt[population]==pop2].index.unique()]
            acs2D = get_acs2D(p_gen_df,p2_gen_df,mode=mode)
            hists2d.update({(pop,pop2):acs2D})
    return hists,hists2d
```

```python
def reduce_pop_hists(pop_hists):
    hists1d = pop_hists[0][0].copy()
    hists2d = pop_hists[0][1].copy()
    for (chrom_hists1d,chrom_hists2d) in pop_hists[1:]:
            hists1d.update({key:(hists1d[key][0]+chrom_hists1d[key][0],hists1d[key][1]) \
                                                for key in hists1d})
            hists2d.update({key:(hists2d[key][0]+chrom_hists2d[key][0],hists2d[key][1],hists2d[key][2]) \
                                                for key in hists2d})
    return hists1d, hists2d
```

### Pairwise differences

#### Total per comparison

```python
%%px --local
def pairwise_diff_numpy(gen_arr):
    """Squared pairwise distances between all 
    columns of 0,1,2 genotype array arr.
    This matrix based function is at least 10 
    times faster than iterating over columns.
    """
    gen_arr = gen_arr.astype(np.float64)-1
    #compare heterozygous with hom alt
    mat1 = np.where(gen_arr==0,-1,gen_arr)
    mat1[np.isnan(gen_arr)]=0
    #and hom ref
    mat2 = np.where(gen_arr==0,1,gen_arr)
    mat2[np.isnan(gen_arr)]=0
    #account for heterozygous comparisons
    mat3 = np.where(gen_arr==0,1,0)
    #don't count nan comparisons
    n = np.dot((~np.isnan(gen_arr)*1.).T,~np.isnan(gen_arr)*1.)
    del gen_arr
    B = (np.dot(mat1.T,mat1)+np.dot(mat2.T,mat2))/2.
    del mat1
    del mat2
    gc.collect()
    het = np.dot(mat3.T,mat3)
    return ((n- B)+het)/2. + np.diag(np.diag(((n- B)+het)/2.)) #self comparisons should not be divided by 2

def pairwise_diff_mat(df):
    """
    Calculate pairwise difference data frame.
    For Genotype 0,1,2 data frame.
    Uses numpy matrix multiplication.
    """
    cols = df.columns
    diff = pairwise_diff_numpy(df.values)
    gc.collect()
    return pd.DataFrame(diff,index=cols,columns=cols)
def get_pw_diff(chrom):
    gen_df = get_gen_df(chrom)
    return pairwise_diff_mat(gen_df)
```

```python
def write_total_pw_diff():
    #attention memory hungy, not more than 8 processes
    pw_aut_map = lv.map_async(get_pw_diff,
                              [i for j in zip(autosomes[:int(len(autosomes)/2)],autosomes[-1:int(-len(autosomes)/2):-1]) for i in j])
    pw_aut_map.wait()
    total_pw_diff = sum(pw_aut_map.result)
    total_pw_diff.to_csv(jn(var_ana_dir,"total_pairwise_diff.tsv"),sep="\t")
    return total_pw_diff
```

```python
%%px --local
def get_fst(chrom,pop0,pop1):
    fn = jn(man_ana_dir, "163_pop3_{}_{}_{}.weir.fst".format(pop0,pop1,chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",squeeze=True,na_values=["nan","-nan"])
    return df

def mean_fst_df(chrom):
    poptuples = [[(pop0,pop1) for pop1 in pops3[i+1:]] for i,pop0 in enumerate(pops3)]
    poptuples = [item for sublist in poptuples for item in sublist]
    index = pd.MultiIndex.from_tuples(poptuples, names=['pop0', 'pop1'])
    out_df = pd.DataFrame(columns=[str(f) for f in np.arange(0.1,0.501,0.1)],index=index)
    n_snp_df = pd.DataFrame(columns=[str(f) for f in np.arange(0.1,0.501,0.1)],index=index)
    gen_df = get_gen_df(chrom)
    for pop01 in poptuples:
        pop0 = pop01[0]
        pop1 = pop01[1]
        fst_df = get_fst(chrom,pop0,pop1)
        gen_df_2pop = gen_df[mt[(mt["pop3"]==pop0)|(mt["pop3"]==pop1)].index]
        af = gen_df_2pop.mean(axis=1)/2.
        maf = (af<0.5)*af + (af>=0.5)*(1-af)
        #remove fixed snps
        fst_df = fst_df.ix[maf[maf>0].index.intersection(fst_df.index)]
        fst_means = []
        n_snps = []
        for val, df in fst_df.groupby(((maf+0.049)*10).round()):
            maf_slice = np.array([val-1,val])/10.
            fst_means.append(df.mean())
            n_snps.append(len(df))
        out_df.ix[(pop0,pop1)] = fst_means
        n_snp_df.ix[(pop0,pop1)] = n_snps
    return out_df,n_snp_df

def mean_fst_df_gt_maf(chrom):
    """
    Similar to above but use all SNPs with AF greater than a cutoff.
    """
    poptuples = [[(pop0,pop1) for pop1 in pops3[i+1:]] for i,pop0 in enumerate(pops3)]
    poptuples = [item for sublist in poptuples for item in sublist]
    index = pd.MultiIndex.from_tuples(poptuples, names=['pop0', 'pop1'])
    out_df = pd.DataFrame(columns=[str(f) for f in np.arange(0.1,0.501,0.1)],index=index)
    n_snp_df = pd.DataFrame(columns=[str(f) for f in np.arange(0.1,0.501,0.1)],index=index)
    gen_df = get_gen_df(chrom)
    for pop01 in poptuples:
        pop0 = pop01[0]
        pop1 = pop01[1]
        fst_df = get_fst(chrom,pop0,pop1)
        gen_df_2pop = gen_df[mt[(mt["pop3"]==pop0)|(mt["pop3"]==pop1)].index]
        af = gen_df_2pop.mean(axis=1)/2.
        maf = (af<0.5)*af + (af>=0.5)*(1-af)
        #remove fixed snps
        fst_df = fst_df.ix[maf[maf>0].index.intersection(fst_df.index)]
        fst_means = []
        n_snps = []
        for val, df in fst_df.groupby(((maf+0.049)*10).round()):
            maf_slice = np.array([val-1,val])/10.
            fst_means.append(df.mean())
            n_snps.append(len(df))
        out_df.ix[(pop0,pop1)] = fst_means
        n_snp_df.ix[(pop0,pop1)] = n_snps
    return out_df,n_snp_df
```

```python
def mul(df1,df2):
    assert list(df1.index) == list(df2.index)
    assert list(df1.columns) == list(df2.columns)
    return pd.DataFrame(df1.values*df2.values,index=df1.index,columns=df1.columns)
def reduce_mean_fst(mean_fsts):
    numerator = reduce(lambda x,y: x+y,[mul(el[0],el[1]) for el in mean_fsts])
    denominator = reduce(lambda x,y:x+y, [el[1] for el in mean_fsts])
    return numerator.div(denominator)
```

```python
def draw_pie(ax,ratios=[0.4,0.3,0.3], X=0, Y=0, size = 1000,colors=colors3):
    N = len(ratios)
 
    xy = []
 
    start = 0.
    for ratio in ratios:
        x = [0] + np.cos(np.linspace(2*math.pi*start,2*math.pi*(start+ratio), 30)).tolist()
        y = [0] + np.sin(np.linspace(2*math.pi*start,2*math.pi*(start+ratio), 30)).tolist()
        xy1 = zip(x,y)
        xy.append(xy1)
        start += ratio
 
    for i, xyi in enumerate(xy):
        ax.scatter([X],[Y] , marker=(xyi,0), s=size, facecolor=colors[i],alpha=0.7)
```

```python
def admix_subplot(data,ax):
    data.plot(kind="bar",stacked=True,ax=ax,legend=False,yticks=[],grid=False)
    ax.set_ylim([0,1])
    #hide y-axis ticks
    ax.get_yaxis().set_visible(False)
    ax2  = ax.twiny()
    tick_pos1 = ax.xaxis.get_majorticklocs()
    #ax2.set_xticks(ax.xaxis.get_majorticklocs())
#    ax2.set_xticklabels(mt.ix[data.index]["pop2"].values)
    #tick_pos2 = 0.05+np.arange(len(data)*1.)/len(data)
    ax2.set_xticks(tick_pos1*1./len(data))
    xlabels = ax2.get_xticklabels() 
    for label in xlabels: 
        label.set_rotation(90) 
    
    return ax,ax2

def admix_plot(data,ind_per_line,K):
    n_subplots = int(np.ceil(len(data)*1./ind_per_line))
    fig = plt.figure(figsize=(24, 2.5*n_subplots))
    #y=0.94 reduces the space between title and plot
    fig.suptitle("admixture K=" + str(K), fontsize=25,y=1.06)
    #increases space between subplots to make space for country labels
    hspace = 1.
    fig.subplots_adjust(hspace=hspace)
    
    bottoms = []
    #print "n_subplots",n_subplots
    for i in range(1,n_subplots+1):
        #print i 
        ax = fig.add_subplot(n_subplots,1,i)
        ax, ax2 = admix_subplot(data.iloc[(i-1)*ind_per_line:i*ind_per_line],ax)
        bottoms.append(ax.get_axes().get_position()._points[0,1])

    final_data = data.iloc[(n_subplots-1)*ind_per_line:]
    pos = ax.get_axes().get_position()
    left = pos._points[0,0]
    width = (pos._points[1,0]- pos._points[0,0])*len(final_data)*1./ind_per_line
    height = pos._points[1,1]- pos._points[0,1]
    try:
        bottom = pos._points[0,1] - (bottoms[-2]- bottoms[-1]) #-hspace
    except IndexError:
        bottom = 0#pos._points[0,1] - (0- bottoms[-1])
    ax = fig.add_subplot(n_subplots,1,n_subplots)
    ax,ax2 = admix_subplot(final_data,ax)
    ax.set_position([left,bottom,width,height])
    ax2.set_position([left,bottom,width,height])
```

### Linkage disequilibrium

```python
%%px --local
def calc_r2(gen_df_pop,snp_distance,tolerance=0.1,rand_deviation=0):
    """
    calculate correlation coefficient between snps
    
    tolerance ... fraction of deviation in distance 
                    that is tolerated (0.1 means +-10%)
    rand_deviation ... add a random deviation so that 
                        not too many neighboring snps have 
                                            the same partner
                        NOT IMPLEMENTED
    """
    chrom = gen_df_pop.index[0][0]
    #snps that have a counter part snp_distance bp further
    last_idx = gen_df_pop.ix[chrom].index.searchsorted(gen_df_pop.ix[chrom].index[-1]-snp_distance)
    idx = gen_df_pop.ix[chrom].index[:last_idx-1]
    # 

    dev_vec = np.random.randint(-1-rand_deviation,1+rand_deviation,len(idx))
    #index snp_distance base-pairs apart
    second_idx = gen_df_pop.ix[chrom].index[idx.searchsorted(idx+snp_distance)+dev_vec]
    real_dist = second_idx.values - idx.values
    #print "total pairs",len(real_dist)
    useable = (real_dist > snp_distance*(1-tolerance))&(real_dist < snp_distance*(1+tolerance))
    idx = idx[useable]
    second_idx = second_idx[useable]
    real_dist = real_dist[useable]
    #print "pairs within tolerance",len(real_dist)
    df1 = gen_df_pop.ix[chrom].ix[idx]
    df2 = gen_df_pop.ix[chrom].ix[second_idx]
    index = df1.index + real_dist/2
    df2.index = index
    df1.index = index + 0
    not_null = df1.notnull().mul(df2.notnull())
    max_missing_frac = 0.4
    not_null = not_null.mul((not_null.mean(axis=1)>(1-max_missing_frac)),axis=0)
    df1 = df1[not_null]
    df2 = df2[not_null]

    covariance = pd.DataFrame(df1.values*df2.values, columns=df1.columns, index=df1.index).mean(axis=1) -\
                        df1.mean(axis=1) *\
                        df2.mean(axis=1)
    def var(df):
        return (df**2).mean(axis=1)- df.mean(axis=1)**2
    
    r2 =  pd.DataFrame({"r2":covariance**2/(var(df1)*var(df2)),"dist":real_dist})
    try:
        r2.index = pd.MultiIndex.from_tuples(zip([chrom]*len(r2.index),r2.index))
    except TypeError:
        print chrom,snp_distance,r2
    return r2
```

#### Get the empirical value of r2 for unlinked sites at hardy weinberg as a function of sample size and allele frequency spectrum

```python
def get_random_genotype_df(gen_df,sfs):
    """
    generate a df with the same index as the input df
    but with allele frequencies randomly drawn from the site frequency spectrum
    for each site genotypes are then drawn assuming hardy weinberg equilibrium
    thus, all snps are unlinked

    sfs: np.histogram result, tuple of counts and bins
    """
    n_sites = gen_df.shape[0]
    af_probas = sfs[0].astype(float)/sum(sfs[0])
    n_chroms = len(af_probas) - 1 
    af_values = (sfs[1][:-1]+0.5)/n_chroms
    af_vec = np.random.choice(af_values, 
                                p=af_probas, 
                                size=n_sites)
    n = n_chroms/2
    assert n == gen_df.shape[1]
    unlinked_gen_array = np.array(map(lambda af: get_random_genotypes(af,n), af_vec))
    return pd.DataFrame(unlinked_gen_array,columns=gen_df.columns,index=gen_df.index)

def get_random_genotypes(af,n):
    return np.random.choice([0,1,2], 
                            p=[(1-af)**2,2*af*(1-af),af**2], 
                            size=n)

#use empirical allele frequency probability
def get_r2_for_unlinked_sites(gen_df,pop):
    subpop_df = get_subpop_df(gen_df,pop)
    unlinked_subpop_df = get_random_genotype_df(subpop_df,hists1d[pop])
    snp_distance = 5000 #choice of this value should not matter
    r2 = calc_r2(unlinked_subpop_df,snp_distance,tolerance=0.1,rand_deviation=0)
    return r2.mean()["r2"]
```

### PCA

#### distance matrix

```python
""" a small class for Principal Component Analysis
Usage:
    p = PCA( A, fraction=0.90 )
In:
    A: an array of e.g. 1000 observations x 20 variables, 1000 rows x 20 columns
    fraction: use principal components that account for e.g.
        90 % of the total variance

Out:
    p.U, p.d, p.Vt: from numpy.linalg.svd, A = U . d . Vt
    p.dinv: 1/d or 0, see NR
    p.eigen: the eigenvalues of A*A, in decreasing order (p.d**2).
        eigen[j] / eigen.sum() is variable j's fraction of the total variance;
        look at the first few eigen[] to see how many PCs get to 90 %, 95 % ...
    p.npc: number of principal components,
        e.g. 2 if the top 2 eigenvalues are >= `fraction` of the total.
        It's ok to change this; methods use the current value.

Methods:
    The methods of class PCA transform vectors or arrays of e.g.
    20 variables, 2 principal components and 1000 observations,
    using partial matrices U' d' Vt', parts of the full U d Vt:
    A ~ U' . d' . Vt' where e.g.
        U' is 1000 x 2
        d' is diag([ d0, d1 ]), the 2 largest singular values
        Vt' is 2 x 20.  Dropping the primes,

    d . Vt      2 principal vars = p.vars_pc( 20 vars )
    U           1000 obs = p.pc_obs( 2 principal vars )
    U . d . Vt  1000 obs, p.obs( 20 vars ) = pc_obs( vars_pc( vars ))
        fast approximate A . vars, using the `npc` principal components

    Ut              2 pcs = p.obs_pc( 1000 obs )
    V . dinv        20 vars = p.pc_vars( 2 principal vars )
    V . dinv . Ut   20 vars, p.vars( 1000 obs ) = pc_vars( obs_pc( obs )),
        fast approximate Ainverse . obs: vars that give ~ those obs.


Notes:
    PCA does not center or scale A; you usually want to first
        A -= A.mean(A, axis=0)
        A /= A.std(A, axis=0)
    with the little class Center or the like, below.

See also:
    http://en.wikipedia.org/wiki/Principal_component_analysis
    http://en.wikipedia.org/wiki/Singular_value_decomposition
    Press et al., Numerical Recipes (2 or 3 ed), SVD
    PCA micro-tutorial
    iris-pca .py .png

"""

from __future__ import division
import numpy as np
dot = np.dot
    # import bz.numpyutil as nu
    # dot = nu.pdot

__version__ = "2010-04-14 apr"
__author_email__ = "denis-bz-py at t-online dot de"

#...............................................................................
class PCA:
    def __init__( self, A, fraction=0.90 ):
        assert 0 <= fraction <= 1
            # A = U . diag(d) . Vt, O( m n^2 ), lapack_lite --
        self.U, self.d, self.Vt = np.linalg.svd( A, full_matrices=False )
        assert np.all( self.d[:-1] >= self.d[1:] )  # sorted
        self.eigen = self.d**2
        self.sumvariance = np.cumsum(self.eigen)
        self.sumvariance /= self.sumvariance[-1]
        self.npc = np.searchsorted( self.sumvariance, fraction ) + 1
        self.dinv = np.array([ 1/d if d > self.d[0] * 1e-6  else 0
                                for d in self.d ])

    def pc( self ):
        """ e.g. 1000 x 2 U[:, :npc] * d[:npc], to plot etc. """
        n = self.npc
        return self.U[:, :n] * self.d[:n]

    # These 1-line methods may not be worth the bother;
    # then use U d Vt directly --

    def vars_pc( self, x ):
        n = self.npc
        return self.d[:n] * dot( self.Vt[:n], x.T ).T  # 20 vars -> 2 principal

    def pc_vars( self, p ):
        n = self.npc
        return dot( self.Vt[:n].T, (self.dinv[:n] * p).T ) .T  # 2 PC -> 20 vars

    def pc_obs( self, p ):
        n = self.npc
        return dot( self.U[:, :n], p.T )  # 2 principal -> 1000 obs

    def obs_pc( self, obs ):
        n = self.npc
        return dot( self.U[:, :n].T, obs ) .T  # 1000 obs -> 2 principal

    def obs( self, x ):
        return self.pc_obs( self.vars_pc(x) )  # 20 vars -> 2 principal -> 1000 obs

    def vars( self, obs ):
        return self.pc_vars( self.obs_pc(obs) )  # 1000 obs -> 2 principal -> 20 vars


class Center:
    """ A -= A.mean() /= A.std(), inplace -- use A.copy() if need be
        uncenter(x) == original A . x
    """
        # mttiw
    def __init__( self, A, axis=0, scale=True, verbose=1 ):
        self.mean = A.mean(axis=axis)
        if verbose:
            print "Center -= A.mean:", self.mean
        A -= self.mean
        if scale:
            std = A.std(axis=axis)
            self.std = np.where( std, std, 1. )
            if verbose:
                print "Center /= A.std:", self.std
            A /= self.std
        else:
            self.std = np.ones( A.shape[-1] )
        self.A = A

    def uncenter( self, x ):
        return np.dot( self.A, x * self.std ) + np.dot( x, self.mean )
```

```python
def plot_genes(gene_df,ax=None,xlim=None):
    if ax is None:
        ax = plt.gca()
        
    gene_df.sort('start', inplace=True)
    bar_thickness = 0.05
    text_space = 0.05
    text_dist = 0.01
    row_height = bar_thickness + text_space
    
    row = 0
    max_row = 0
    prev_end = -np.inf
    for i, g in gene_df.iterrows():
        if g['start'] <= prev_end:
            row += 1
        else:
            row = 0
        ax.annotate(g['symbol'], ((g['start']+g['end'])/2.,-bar_thickness-row*row_height-text_dist), #
                      size=10, ha="right", va="top",rotation=30)#rotation = 270,+g['length']/2.
        plt.barh(bottom= -bar_thickness-row*row_height,
                 left=g['start'],
                 width=g['end']-g['start'],
                 height=bar_thickness,
                 color = 'yellow', alpha=0.5) 
        max_row = max(max_row, row)
        prev_end = g['end']
        
    
           
  
    ax.set_ylim([-(max_row+1)*row_height,0])
    
    if xlim is None:
        ax.set_xlim([gene_df['start'].min(),gene_df['end'].max()])
    else:
        ax.set_xlim(xlim)
    ax.set_axis_off()
    
    return ax
```

```python
%%px --local
def get_window_pi(chrom,pop):
    window_pi = pd.read_csv(jn(man_ana_dir,"{}_{}.windowed.pi".format(pop,chrom)), sep='\t')
    window_pi.columns = [s.lower() for s in window_pi.columns]
    assert len((window_pi['bin_end'] - window_pi['bin_start']).unique()) == 1, "Windows of unequal lenght not supported"
    window = (window_pi['bin_end'] - window_pi['bin_start']).iloc[0]
    window_pi['pos'] = (window_pi['bin_start'] + window/2.).astype(int)
    s = window_pi.set_index(['chrom','pos'])['pi']
    s.name = 'pi_{}'.format(window)
    return s
def get_window_tajd(chrom,pop,window):
    window_tajD = pd.read_csv(jn(man_ana_dir,"{}_{}_window{}.Tajima.D".format(pop,chrom,window)), sep='\t')
    window_tajD.columns = [s.lower() for s in window_tajD.columns]
    window_tajD['pos'] = (window_tajD['bin_start'] + window/2.).astype(int)
    s = window_tajD.set_index(['chrom','pos'])['tajimad']
    s.name = 'TajimaD_{}'.format(window)
    return s
def get_fst(chrom,pop0,pop1):
    fn = jn(man_ana_dir, "163_pop3_{}_{}_{}.weir.fst".format(pop0,pop1,chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",squeeze=True,na_values=["nan","-nan"])
    return df
```

```python
%%px --local
def get_window_pi_all_chrom(pop):
    s_all_chrom = pd.DataFrame()
    for chrom in chromosomes[:-1]:
        s = get_window_pi(chrom,pop)
        s_all_chrom = pd.concat([s_all_chrom,s])
    #get a series instead of df
    s_all_chrom = s_all_chrom.iloc[:,0]
    s_all_chrom.index = pd.MultiIndex.from_tuples(s_all_chrom.index)
    
    s_all_chrom.name = s.name
    s_all_chrom.index.names = ['chrom','pos']
    #return 'bla'
    return s_all_chrom
def get_tajd_all_chrom(pop,window):
    s_all_chrom = pd.DataFrame()
    for chrom in chromosomes[:-1]:
        s = get_window_tajd(chrom,pop,window)
        s_all_chrom = pd.concat([s_all_chrom,s])
    #get a series instead of df
    s_all_chrom = s_all_chrom.iloc[:,0]
    s_all_chrom.index = pd.MultiIndex.from_tuples(s_all_chrom.index)
    s_all_chrom.name = s.name
    s_all_chrom.index.names = ['chrom','pos']
    #return 'bla'
    return s_all_chrom
    
```

### Balancing selection

#### Number of populations a SNP is seggregating in

```python
%%px --local
def get_shared_snps(chrom,poptype,af_thresh):
    """
    all snps are considered
    af_thresh ... in sub populations
    """
    def af(df):
        return df.mean(axis=1)/2.
    gen_df = get_gen_df(chrom)
    pop3_af = gen_df.groupby(mt[poptype],axis=1).apply(af)
    # for each snp, count in how many pops it it seggretating
    n_pop = ((pop3_af>af_thresh)&(pop3_af<1-af_thresh)).sum(axis=1)
    n_pop.name = "n_pops"
    return n_pop

def get_shared_snps_global_thresh(chrom,poptype,global_af_thresh,local_af_thresh):
    """
    only consider SNPs with af>global_af_thresh
    """
    def af(df):
        return df.mean(axis=1)/2.
    gen_df = get_gen_df(chrom)
    gaf = af(gen_df)
    gen_df = gen_df[(global_af_thresh<gaf)&(gaf<(1-global_af_thresh))]
    pop3_af = gen_df.groupby(mt[poptype],axis=1).apply(af)
    # for each snp, count in how many pops it it seggretating
    n_pop = ((pop3_af>local_af_thresh)&(pop3_af<1-local_af_thresh)).sum(axis=1)
    n_pop.name = "n_pops"
    return n_pop

def get_shared_snps_window(chrom,poptype,af_thresh, window, overlap, min_n_values):
    n_pop_s = get_shared_snps(chrom, poptype=poptype, af_thresh=af_thresh)
    window_npops = hpd.index_rolling(n_pop_s.ix[chrom], window,lambda s: s.mean(), 
                                            overlap=overlap, min_n_values=min_n_values)
    window_npops.index = pd.MultiIndex.from_arrays([[chrom]*len(window_npops),window_npops.index])
    window_npops.name = "n_pop"
    window_npops.index.names = ["chrom","pos"]
    return window_npops
```

```python
def write_ballet_input(chrom):
    """
    Write BALLET input files according to
    http://www.personal.psu.edu/mxd60/Manual_BALLET_v1.0.pdf
    
    (M DeGiorgio, KE Lohmueller, R Nielsen (2014))
    """
    def get_poly(df):
        ac = df.sum(axis=1)
        n = 2*df.notnull().sum(axis=1)
        df = pd.DataFrame({'x':n.values-ac.values,'n':n.values}, index=ac.index.droplevel(0))
        df = df[['x','n']]
        df.index.name = 'position'
        df.dropna(inplace=True)
        return df[df['x']!=df['n']].astype(int)
    ad_df = get_ad_df(chrom)
    popgroup = ad_df.groupby(lambda id: mt['pop3'].ix[id],axis=1)
    for pop, df in popgroup:
        poly_df = get_poly(df)
        
        poly_df.to_csv(jn(ballet_dir,"163_UG_ballet_polysub_{}_{}.tsv".format(pop,chrom)),sep='\t')
        
        pos_s = pd.Series(np.nan,index=poly_df.index)
        pos_s.name = 'centimorgan'
        map_values = map_s.ix[chrom].add(pos_s,fill_value=0).interpolate(method='values')
        
        #interpolating the first values
        first_idx, second_idx = map_s.ix[chrom].head(2).index.values
        first_value, second_value = map_s.ix[chrom].head(2).values
        map_values.ix[:first_idx] = (second_value - first_value) / (second_idx - first_idx) * \
                                                    map_values.ix[:first_idx].index + \
                                                    (first_value *second_idx - second_value * first_idx) / (second_idx - first_idx)
        #interpolating the last values
        secondlast_idx, last_idx = map_s.ix[chrom].tail(2).index.values
        secondlast_value, last_value = map_s.ix[chrom].tail(2).values
        map_values.ix[last_idx:] = (last_value - secondlast_value) / (last_idx - secondlast_idx) * \
                                                    map_values.ix[last_idx:].index + \
                                                    (secondlast_value *last_idx - last_value * secondlast_idx) / (last_idx - secondlast_idx)
        rho_s = map_values.ix[pos_s.index].diff().fillna(0)*(4*neff[pop])/100.
        rho_s.name = 'rate'
        
        rho_s.to_csv(jn(ballet_dir,"163_UG_ballet_rec_{}_{}.tsv".format(pop,chrom)),sep='\t',header=True)
```

#### Interpolate genetic map

```python
%%px --local
def make_map_df(chrom):
    gen_df = get_gen_df(chrom)
    idx = gen_df.ix[chrom].index.union(map_s.ix[chrom].index)
    idx.name = "basepair"
    genetic_map = pd.DataFrame(index=idx)
    genetic_map["centimorgan"] = np.nan
    genetic_map.ix[map_s.ix[chrom].index,"centimorgan"] = map_s.ix[chrom].values
    genetic_map["163_pop_snp"] = 0
    genetic_map.ix[gen_df.ix[chrom].index,"163_pop_snp"] = 1
    genetic_map["vrc_snp"] = 0
    genetic_map.ix[map_s.ix[chrom].index,"vrc_snp"] = 1
    genetic_map["centimorgan"] = genetic_map["centimorgan"].interpolate(method='values')
    #interpolating the first values
    first_idx, second_idx = map_s.ix[chrom].head(2).index.values
    first_value, second_value = map_s.ix[chrom].head(2).values
    genetic_map.ix[:first_idx,"centimorgan"] = (second_value - first_value) / (second_idx - first_idx) * \
                                                genetic_map.ix[:first_idx,"centimorgan"].index + \
                                                (first_value *second_idx - second_value * first_idx) / (second_idx - first_idx)
    #interpolating the last values
    secondlast_idx, last_idx = map_s.ix[chrom].tail(2).index.values
    secondlast_value, last_value = map_s.ix[chrom].tail(2).values
    genetic_map.ix[last_idx:,"centimorgan"] = (last_value - secondlast_value) / (last_idx - secondlast_idx) * \
                                                genetic_map.ix[last_idx:,"centimorgan"].index + \
                                                (secondlast_value *last_idx - last_value * secondlast_idx) / (last_idx - secondlast_idx)
    genetic_map.index = pd.MultiIndex.from_tuples(zip([chrom]*len(idx),idx))
    genetic_map.index.names = ["chrom","pos"]
    return genetic_map

def save_genetic_map(gen_map_df):
    chrom = gen_map_df.index.levels[0][0]
    fn = jn(var_ana_dir,"163_UG_ref3500_genetic_map_{}.tsv".format(chrom))
    gen_map_df.index.names = ["chrom","pos"]
    gen_map_df.to_csv(fn, sep="\t",float_format="%.5g")
    
def save_genetic_map_chrom(chrom):
    map_df = make_map_df(chrom)
    save_genetic_map(map_df)
```

```python
def reduce_dic(dic_ls):
    """
    Takes list of dictionaries with identical keys
    and adds up the values for each key.
    """
    dic = dic_ls[0].copy()
    for k in dic:
        dic[k] = dic[k] + sum([d[k] for d in dic_ls[1:]])
    return dic
```

```python
%%px --local
def get_filter_stats(chrom):
    filter_dic = json.load(open(jn(var_ana_dir,"filter_stats_{}.json".format(chrom))))
    assert filter_dic["total"] == filter_dic["filter_nosnp"]+filter_dic["filter_snp"]+filter_dic["pass_nosnp"]+filter_dic["pass_snp"]
    return filter_dic

def get_N_stats(chrom):
    N_stats = pd.read_csv(jn(var_ana_dir,"N_stats_{}.tsv".format(chrom)),sep="\t",index_col=0)
    return N_stats

def get_rel_N_stats(N_stats,filter_dic):
    rel_N_stats = N_stats.copy()
    for col in N_stats:
        rel_N_stats[col] = N_stats[col] *1. / filter_dic[col]
    rel_N_stats.columns = ["N_"+r for r in rel_N_stats.columns]
    rel_N_stats["cov"] = cov_df.ix["mean_autosome"]
    #per individual sites where I can't say anything (either filtered or N)
    rel_N_stats["filter_or_N_snp"] =  (N_stats["pass_snp"]+filter_dic["filter_snp"])*1./\
                                        (filter_dic["filter_snp"]+filter_dic["pass_snp"])
    rel_N_stats["filter_or_N_nosnp"] =  (N_stats["pass_nosnp"]+filter_dic["filter_nosnp"])*1./\
                                            (filter_dic["filter_nosnp"]+filter_dic["pass_nosnp"])
    rel_N_stats["filter_or_N_overall"] = (N_stats["pass_nosnp"]+N_stats["pass_snp"]+filter_dic["filter_nosnp"]+filter_dic["filter_snp"])*1.\
                                            /(filter_dic["total"])
    return rel_N_stats

def get_stats(chrom):
    fs = get_filter_stats(chrom)
    ns =  get_N_stats(chrom)
    return fs, ns

def reduce_stats(stat_ls):
    return reduce_dic([el[0] for el in stat_ls]),\
            sum([el[1] for el in stat_ls])
    
def get_N_corr(chrom):
    return pd.read_csv(jn(var_ana_dir,"N_corr_{}.tsv".format(chrom)),sep="\t",index_col=0)
```

```python
def write_tot_stats():
    #reduce total filter and N_stats
    stats = lv.map_async(get_stats,autosomes)
    stats.wait()
    tot_site_stats, tot_N_stats = reduce_stats(stats.result)
    json.dump(tot_site_stats,open(jn(var_ana_dir,"tot_site_stats.json"),"w"))
    tot_N_stats.to_csv(jn(var_ana_dir,"tot_N_stats.tsv"),sep="\t")
def get_tot_stats():
    return json.load(open(jn(var_ana_dir,"tot_site_stats.json"))),\
            pd.read_csv(jn(var_ana_dir,"tot_N_stats.tsv"),sep="\t",index_col=0)
```

```python
%%px --local
validation_dir = eu("~/vervet_project/analyses/20140617_ref3500_pacbio_validation/_data")
individuals = ['AG5417','B5616','AGM141','VEC1016','VGA00006','VZA3008','VSAM2001','VGA00146']

def get_pacbio_intervals(ind):
    intervals = pd.read_csv(jn(validation_dir,"pacbio_mapping_intervals_{}.bed".format(ind)),
                               sep="\t",
                                header=None,index_col=[0,1],
                                names=["chrom","pos","end"],usecols=[0,1,2])
    intervals["start"] =intervals.index.droplevel(0)
    intervals.index = intervals.index.droplevel(1)
    return intervals
    

vcf_fn = jn(var_ana_dir,"163_UG_ref3500_all_sites_{}_filter_anc.vcf.gz")
pacbio_fn = jn(validation_dir,
              "pacbio_samtools_all_sites_{}_filter.vcf.gz")
```

```python
%%px --local
def get_validation_countdic(chrom,ind,check_fun):
    """
    check_fun(pl,il,count_dic,gt) ... 
                function that checks 
                checks the site and adds the count to countdic
    """
    intervals = get_pacbio_intervals(ind)
    pb = tabix.open(pacbio_fn.format(ind))
    ill = tabix.open(vcf_fn.format(chrom))
    for line in gzip.open(vcf_fn.format(chrom)):
        if line[0] == '#':
            if line[1:1+len('CHROM')] == 'CHROM':
                ill_header = line[1:].strip().split('\t')
        else:
            break
    ind_pos = ill_header.index(ind)
    count_dic = {'total':0,'both_filter':0,'pb_filter':0,'ill_filter':0,'ill_N':0,
                 'pb_ref': {'ill_RR':0,'ill_RA':0,'ill_AA':0},
                 'pb_noref':{'ill_RR':0,'ill_RA':0,'ill_AA':0},
                  'other':0}

    ind_pos = ill_header.index(ind)
    try:
        iv_df = intervals.ix[chrom].copy()
        #make sure that intervals don't overlap
        iv_df['start'].iloc[1:] =  np.array([iv_df['end'].iloc[:-1].values.T,iv_df['start'].iloc[1:].values.T]).T.max(axis=1)
        iv_df = iv_df[iv_df['start']<iv_df['end']]
        iv_itr = iv_df.iterrows()
    except KeyError:
        return {(ind,chrom):count_dic}
    except AttributeError:
        iv_itr = pd.DataFrame(intervals.ix[chrom],columns=[chrom]).T.iterrows()
    for i,(c,iv) in enumerate(iv_itr):
        try:
            pbr = pb.query(c,iv['start']+1,iv['end'])
        except tabix.TabixError:
            return (c,iv['start']+1,iv['end'])
        illr = ill.query(c,iv['start']+1,iv['end'])
        for j,(pl,il) in enumerate(itertools.izip(pbr,illr)):
            assert pl[0] == il[0]
            if iv['end'] ==  17423957:
                print 'pac:',pl[1], 'ill:', il[1]
            while int(pl[1]) != int(il[1]):
                if int(pl[1]) < int(il[1]):
                    try:
                        pl = pbr.next()
                    except StopIteration:
                        break
                elif int(pl[1]) > int(il[1]):
                    try:
                        il = illr.next()
                    except StopIteration:
                        break
            if int(pl[1]) > iv['end'] or int(il[1]) > iv['end']:
                break
            assert pl[1] == il[1], ind + " " + c + " " + str(iv) + ": " + " ".join(pl) + '\n' + " ".join(il)
            count_dic['total'] += 1
            gt = il[ind_pos][:3]
            check_fun(pl,il,count_dic,gt)
    return {(ind,chrom):count_dic}
```

```python
%%px --local
nofilter = ['.','PASS']
gts = ['0/0','0/1','1/1']
def check_filter(pl,il,count_dic,gt):
    if pl[6] not in nofilter and il[6] not in nofilter:
        count_dic['both_filter'] += 1
    elif pl[6] not in nofilter:
        count_dic['pb_filter'] += 1
    elif il[6] not in nofilter:
        count_dic['ill_filter'] += 1
    elif gt not in gts:
        count_dic['ill_N'] += 1
    elif pl[4] == 'X' and gt == gts[0]:
        count_dic['pb_ref']['ill_RR'] += 1
    elif pl[4] == 'X' and gt == gts[1]:
        count_dic['pb_ref']['ill_RA'] += 1
    elif pl[4] == 'X' and gt == gts[2]:
        count_dic['pb_ref']['ill_AA'] += 1
    elif pl[4] != 'X' and gt == gts[0]:
        count_dic['pb_noref']['ill_RR'] += 1
    elif pl[4] != 'X' and gt == gts[1]:
        count_dic['pb_noref']['ill_RA'] += 1
    elif pl[4] != 'X' and gt == gts[2]:
        count_dic['pb_noref']['ill_AA'] += 1
    else:
        count_dic['other'] += 1
        
def check_filter_Q1(pl,il,count_dic,gt):
    if pl[6] not in nofilter and il[6] not in nofilter:
        count_dic['both_filter'] += 1
    elif pl[6] not in nofilter:
        count_dic['pb_filter'] += 1
    elif il[6] not in nofilter:
        count_dic['ill_filter'] += 1
    elif gt not in gts:
        count_dic['ill_N'] += 1
    elif pl[4] == 'X' and gt == gts[0]:
        count_dic['pb_ref']['ill_RR'] += 1
    elif pl[4] == 'X' and gt == gts[1]:
        count_dic['pb_ref']['ill_RA'] += 1
    elif pl[4] == 'X' and gt == gts[2]:
        count_dic['pb_ref']['ill_AA'] += 1
    elif pl[4] != 'X' and float(pl[5])<1:
        count_dic['pb_filter'] += 1
    elif pl[4] != 'X' and gt == gts[0]:
        count_dic['pb_noref']['ill_RR'] += 1
    elif pl[4] != 'X' and gt == gts[1]:
        count_dic['pb_noref']['ill_RA'] += 1
    elif pl[4] != 'X' and gt == gts[2]:
        count_dic['pb_noref']['ill_AA'] += 1
    else:
        count_dic['other'] += 1
        
def check_Q1_only(pl,il,count_dic,gt):
    """
    we ignore the countdic entry both filter here
    and attribute everything that is filtered in illumina
    directly to 'ill_filter'
    """
    if il[6] not in nofilter:
        count_dic['ill_filter'] += 1
    elif gt not in gts:
        count_dic['ill_N'] += 1
    elif pl[4] == 'X' and gt == gts[0]:
        count_dic['pb_ref']['ill_RR'] += 1
    elif pl[4] == 'X' and gt == gts[1]:
        count_dic['pb_ref']['ill_RA'] += 1
    elif pl[4] == 'X' and gt == gts[2]:
        count_dic['pb_ref']['ill_AA'] += 1
    elif pl[4] != 'X' and float(pl[5])<1:
        count_dic['pb_filter'] += 1
    elif pl[4] != 'X' and gt == gts[0]:
        count_dic['pb_noref']['ill_RR'] += 1
    elif pl[4] != 'X' and gt == gts[1]:
        count_dic['pb_noref']['ill_RA'] += 1
    elif pl[4] != 'X' and gt == gts[2]:
        count_dic['pb_noref']['ill_AA'] += 1
    else:
        count_dic['other'] += 1
```

```python
def sum_countdic(c0,c1):
    """
    Add count-dictionaries 
    assuming that they have the same keys.
    Alows them to contain sub-countdictionaries
    E.g. {'total':0,'variant':{'snp':0,'indel':0}}
    """
    c = c0.copy()
    assert set(c0.keys()) == set(c1.keys()), str(c0)+str(c1)
    for k in c0:
        try:
            c[k] += c1[k]
        except TypeError:
            c[k] = sum_countdic(c[k],c1[k])
    return c
def check_total(countdic):
    cntdc=countdic.copy()
    def add_up(cdic):
        t = 0
        for v in cdic.values():
            try:
                t += v
            except TypeError:
                t += add_up(v)
        return t
    total = cntdc['total']
    del cntdc['total']
    count = add_up(cntdc)
    if total != count:
        print "{} != {}".format(total,count)
        return False
    else:
        return True
```

```python
def reduce_map_valid(res):
    return reduce(sum_countdic,[r.values()[0] for r in res])    
def get_valid_pct(valid_dic):
    false_neg = valid_dic['pb_noref']['ill_RR']*1./(sum(valid_dic['pb_noref'].values()))
    false_AA = valid_dic['pb_ref']['ill_AA']*1./(sum(valid_dic['pb_ref'].values()))
    false_RA = (valid_dic['pb_ref']['ill_RA']-valid_dic['pb_noref']['ill_RA'])*1./(valid_dic['pb_ref']['ill_RA']+valid_dic['pb_noref']['ill_RA'])
    return pd.Series({'false_neg':false_neg,'false_AA':false_AA,'false_RA':false_RA,
                     'ill_N':valid_dic['ill_N']*1./valid_dic['total'],
                     'both_filter':valid_dic['both_filter']*1./valid_dic['total'],
                     'ill_filter':valid_dic['ill_filter']*1./valid_dic['total'],
                     'pb_filter':valid_dic['pb_filter']*1./valid_dic['total']})
```

```python
#%%script --bg python
valid_pct_df = pd.DataFrame()
valid_count_df = pd.DataFrame()
for ind in individuals:
    dview['ind'] = ind
    #dview['check_filter_Q1'] = check_filter_Q1
    map_valid = lv.map_async(lambda chrom: get_validation_countdic(chrom,ind,check_Q1_only),chromosomes)
    valid_dic = reduce_map_valid(map_valid.result)
    print ind, check_total(valid_dic)
    json.dump(valid_dic,open(jn(validation_dir,"validation_{}_Q1_only_UG.json".format(ind)),'w'))
    s0 = pd.Series(valid_dic)
    s0.name = ind
    valid_count_df = valid_count_df.append(s0)
    s = get_valid_pct(valid_dic)
    s.name = ind
    valid_pct_df = valid_pct_df.append(s)
valid_pct_df.to_csv(jn(validation_dir,"validation_163_UG_vs_pacbio_pct_Q1_only.tsv"),sep='\t')
valid_count_df.to_csv(jn(validation_dir,"validation_163_UG_vs_pacbio_count_Q1_only.tsv"),sep='\t')
rdme =\
"""
Comparing 2 sequencing types to infer false positve/negative rates.
----------
Data sets:
1) Illumina 100bp paired end at 10X coverage (for the ind. in question)
   + alignment bwa-mem and SNP calling with UnifiedGenotyper and hard filtering
2) pacbio CSS reads with phred >40 of 8 individuals, spanning a small proportion of the genome
   supposedly high quality.
----------
PACBIO filters:
    for SNPs:
        QUALITY >1: this gets rid of most of the indel artefacts
    
CAVEATS: 
    The supposedly high quality circular consens pacbio reads show lots of 1bp indels.
    In the viciniy of these indels there are SNPs which are almost never seen in illumina.
 Hence, the pacbio SNPs should not be taken at face value either, so the estimates here
 are conservative.
EXPLANATION of TABLE CONTENTS:
    false_neg = (sites where pacbio SNP && illumina homozygous ref (RR))/(sites where pacbio SNP)
    false_AA = (sites where pacbio non-SNP && illumina homozygous alt (RR))/(sites where pacbio non-SNP)
    false_RA = (sites where pacbio non-SNP && illumina het (RA) - sites where pacbio SNP && illumina het (RA))
            /(sites where pacbio non-SNP && illumina het (RA) + sites where pacbio SNP && illumina het (RA))
    The idea for false_RA is that, since we only have one haplotype, if a site really is a heterozygote it
    should be seen in pacbio 50% of the time as SNP and 50% as non-SNP. However, false positives in illumina
    (sites that really are RR) should always show up as non-SNPs in pacbio. Note that this, calcuation ignores 
    the fact that some of the sites that are called RA in illumina might really be AA, in which case such 
    a site should always be called SNP in pacbio.
       """
with open(jn(validation_dir,"validation_163_UG_vs_pacbio_pct_Q1_only.README"),'w') as rf:
    rf.write(rdme)
```
