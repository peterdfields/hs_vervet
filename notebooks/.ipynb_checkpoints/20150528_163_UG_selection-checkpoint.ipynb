## prepare XPCLR input files

```python
%%px --local
def get_gen_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)
def get_allele_df(chrom):
    allele_df = pd.read_csv(jn(var_ana_dir,"163_UG_ref3500_{}_pass_ancestral_allele.tsv".format(chrom)),
                            sep="\t",
                            index_col=[0,1],na_values=["NA"])
    return allele_df
def get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac):
    """
    get data-frames for two sub-populations were SNPs not seggregating in at least one of the two are removed
    """
    gen_df_pop01 = gen_df[meta_df[(meta_df["pop3"]==pop0)].index.union(meta_df[(meta_df["pop3"]==pop1)].index)]
    #print len(gen_df_pop)
    #removing non-seggretating sites
    gen_df_pop01 = gen_df_pop01[((gen_df_pop01.mean(axis=1)>0)&(gen_df_pop01.mean(axis=1)<2))]
    gen_df_pop0 = gen_df_pop01[meta_df[(meta_df["pop3"]==pop0)].index]
    gen_df_pop1 = gen_df_pop01[meta_df[(meta_df["pop3"]==pop1)].index]
    not_excess_missing = (gen_df_pop0.isnull().mean(axis=1)<max_missing_frac)&\
                            (gen_df_pop1.isnull().mean(axis=1)<max_missing_frac)
    
    #print len(gen_df_pop)
    return (gen_df_pop0[not_excess_missing],gen_df_pop1[not_excess_missing])

def gen_df_to_019(gen_df):
    df_0 = (gen_df > 0) + gen_df*0 #last term propagates nan
    df_0 = df_0.fillna(9)
    df_1 = (gen_df > 1) + gen_df*0
    df_1 = df_1.fillna(9)
    #to shuffle the haplotypes randomly
    shuffle_mat = np.random.randint(0,2,df_0.shape)
    df_0s = df_0 * shuffle_mat + df_1 * (1-shuffle_mat)
    df_1s = df_0 * (1-shuffle_mat) + df_1 * shuffle_mat
    assert (df_0s + df_1s == df_0 + df_1).all().all(), "the shuffling goes wrong"
    df_0s.columns = pd.MultiIndex.from_tuples(zip(df_0s.columns,[0]*len(df_0s.columns)))
    df_1s.columns = pd.MultiIndex.from_tuples(zip(df_1s.columns,[1]*len(df_1s.columns)))
    df01 = df_0s.join(df_1s).sort(axis=1).astype(np.int8)
    return df01
def save_geno_file(df_019,fn_base):
    #write header file
    with open(jn(xpclr_ana_dir,fn_base+".individuals"),"w") as f:
        f.write("\t".join([i for i in df_019.columns.droplevel(1).values])+"\n")
    df_019.to_csv(jn(xpclr_ana_dir,fn_base+".geno"),sep=" ",index=False,header=False)
    
def make_snp_file(df_019,allele_df,gen_map_df):
    snp_df = pd.DataFrame({"name":[i[0]+"_"+str(i[1]) for i in df_019.index.values]},index=df_019.index)
    snp_df["chrom"] = map(lambda s: s.split("E")[1],df_019.index.droplevel(1))
    snp_df["morgan"] = gen_map_df.ix[df_019.index]["centimorgan"]/100.
    snp_df["bp"] = df_019.index.droplevel(0)
    snp_df[["ref","alt"]] = allele_df.ix[df_019.index][["REF","ALT"]]
    return snp_df

def save_snp_file(snp_df,fn_base):
    snp_df.to_csv(jn(jn(xpclr_ana_dir,fn_base+".snp")),sep=" ",index=False,header=False)
    
def create_xpclr_input_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac):
    dfs = get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac)
    pops = [pop0, pop1]
    for i,j in [(1,0),(0,1)]:
        df_019 = gen_df_to_019(dfs[i])
        fn_base = "163_xpclr_{}_vs_{}_{}".format(pops[i],pops[j],chrom)
        save_geno_file(df_019,fn_base)
    snp_file0 = make_snp_file(df_019,allele_df,genetic_map)
    save_snp_file(snp_file0,fn_base)

def create_snp_file_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac):
    #only creates the snp file
    dfs = get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac)
    snp_file0 = make_snp_file(dfs,allele_df,genetic_map)
    save_snp_file(snp_file0,fn_base)    
def create_snp_files(chrom,max_missing_frac):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    for i,pop0 in enumerate(pops):
        for pop1 in pops[i+1:]:
            create_snp_file_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac)
    
def create_xpclr_input(chrom):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    for i,pop0 in enumerate(pops):
        for pop1 in pops[i+1:]:
            create_xpclr_input_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map)
            
def write_slice_xpclr_chrom(chrom,slice_size,overlap,max_missing_frac):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    #out = []
    for i, pop0 in enumerate(pops3):
        for pop1 in pops3[i+1:]: 
            df0, df1 = get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac=max_missing_frac)
            df0_019 = gen_df_to_019(df0)
            df1_019 = gen_df_to_019(df1)
            snp_df = make_snp_file(df0_019,allele_df,genetic_map)
            #out.append(
            write_xpclr_in_slices(chrom,pop0,pop1,df0_019,df1_019,snp_df,slice_size,overlap)
            #)
    #return out

def write_xpclr_whole_chrom(chrom,max_missing_frac):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    #out = []
    for i, pop0 in enumerate(pops3):
        for pop1 in pops3[i+1:]: 
            create_xpclr_input_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac)   
    
import json
def write_xpclr_in_slices(chrom,pop0,pop1,df0_019,df1_019,snp_df,slice_size,overlap):
    extensions = ["geno","geno","snp"]
    pops = [(pop0,pop1),(pop1,pop0),(pop0,pop1)]
    for j,df in enumerate([df0_019,df1_019,snp_df]):
        #if j<2:
        #    continue
        fn_base = "163_xpclr_{}_vs_{}_{}".format(pops[j][0],pops[j][1],chrom)
        if j<2:
            with open(jn(xpclr_ana_dir,fn_base+".individuals"),"w") as f:
                f.write("\t".join([str(i) for i in df.columns.droplevel(1).values])+"\n")
        for i,start in enumerate(range(df.head(1).index.droplevel(0)[0],
                           df.tail(1).index.droplevel(0)[0],
                           slice_size-overlap)):
            end = min(start + slice_size, df.ix[chrom].index[-1])
            #try:
            df.ix[chrom].ix[start:end].to_csv(jn(xpclr_ana_dir,fn_base+"_{}".format(i)+"."+extensions[j]),
                                               sep=" ",header=False,index=False,float_format="%.7g")
            #except:
            #    return (df, chrom, start, end)
    info_dic = {"slice_size":slice_size,"overlap":overlap,"n_slice":i+1,"info":"Input files for XPCLR with given slice size and overlap."}
    json.dump(info_dic,open(jn(xpclr_ana_dir,fn_base+".INFO.json"),"w"))           
```

```python
%%px --local
def get_xpclr(chrom,pop0,pop1,i):
    try:
        xpclr_df = pd.read_csv(jn(xpclr_ana_dir,"163_xpclr_{}_vs_{}_{}_{}.result.xpclr.txt".format(pop0,pop1,chrom,i)),
                           header=None,names=["chrom","grid","n_snp","pos","morgan","score","max_s"],sep=" ")
    except IOError:
        with open(jn(xpclr_ana_dir,"missing_files.tsv"),'a') as f:
            f.write(chrom+'\t'+pop0+'\t'+pop1+'\t'+str(i)+'\n')
        return pd.DataFrame()
    xpclr_df["chrom"] = chrom
    xpclr_df.set_index(["chrom","pos"],inplace=True)
    return xpclr_df

def get_xpclr_chrom(chrom,pop0,pop1):
    info = get_info(chrom,pop0,pop1)
    df0 = get_xpclr(chrom,pop0,pop1,0)
    try:
        gridsize = int(round((df0.index[-1][1]-df0.index[0][1])/len(df0)))
    except IndexError:
        gridsize = np.inf
    border = int(info["overlap"] / (2*gridsize))
    df = df0.iloc[:-(border+1)]
    for i in range(1,info["n_slice"]):
        df = pd.concat([df,get_xpclr(chrom,pop0,pop1,i).iloc[border:-(border+1)]]) 
    return df

def get_info(chrom,pop0,pop1):
    popsl = [pop0,pop1]
    popsl.sort()
    return json.load(open(jn(xpclr_ana_dir,"163_xpclr_{}_vs_{}_{}.INFO.json".format(popsl[0],popsl[1],chrom,i))))
 

def xpclr_all_chrom(pop0,pop1):
    return pd.concat([get_xpclr_chrom(chrom,pop0,pop1) for chrom in autosomes])

def save_xpclr_all_chrom(pop0,pop1):
    df = pd.concat([get_xpclr_chrom(chrom,pop0,pop1) for chrom in autosomes])
    df.to_csv(jn(xpclr_ana_dir,"xpclr_all_chrom_{}_vs_{}.tsv".format(pop0,pop1)),sep="\t")

#201510 combine multiple comparisons to compound score 
    
```

```python
def plot_along_chrom(df,chrom,quantile,window_size):
    chr_len = chrom_length.ix[chrom]
    n_subplot = len(pops)
    #print round(window_size/1000)
    fig = plt.figure(1, figsize=(15./(1-log(10000./window_size))\
                           *chr_len/chrom_length.ix["CAE1"],1.5*n_subplot))
    plt.suptitle("Akey's d, "+str(int(window_size/1000))+"kb windows",fontsize="15") 
    hspace = 0.3
    fig.subplots_adjust(hspace=hspace)
    for i,pop in enumerate(pops):
        qt = quantile[pop]
        ax = fig.add_subplot(n_subplot,1,i+1)
        ax.set_frame_on(False)
        ax.set_yticks([0,round(qt)])
        if i < len(pops)-1:
            ax.set_xticks([])
            ax.get_xaxis().set_visible(False)
        else:
            ax.set_xlabel(chrom)
        #total_3window_d[idx][pop].plot(style=".",markersize=5,color=colors[i],ax=ax)
        ax.plot(window_d.ix[chrom][pop].index,window_d.ix[chrom][pop].values,".",markersize=5,color=colors[i])
        ax.set_title(pop,position=(0.5,0.95))
        ax.set_xlim([0,chr_len+1000])
    return fig

def get_chrom_grid(chrom_lens=None,parent_gridelement=None):
    """
    subplot_spec is the parent gird
    """
    if chrom_lens is None:
        #only autosomes
        chrom_lens = np.array([126035930,  90373283,  92142175,  91010382,  75399963,  50890351,
       135778131, 139301422, 125710982, 128595539, 128539186, 108555830,
        98384682, 107702431,  91754291,  75148670,  71996105,  72318688,
        33263144, 130588469, 127223203, 101219884,  82825804,  84932903,
        85787240,  58131712,  48547382,  21531802,  24206276, 130038232,
         6181219])[:-2]
    if parent_gridelement is None:
        parent_gridelement = mpl.gridspec.GridSpec(1, 1)[0] 
    gs = mpl.gridspec.GridSpecFromSubplotSpec(1, len(chrom_lens),
                                                     width_ratios=chrom_lens.astype(float)/sum(chrom_lens),
                                                                                 subplot_spec=parent_gridelement)
    return gs, parent_gridelement

def get_chrom_axes(grid=None,parent_gridelement=None,fig=None,
                      ylabel = "", plot_xlabels=True, color=None):
    """
    axes of the returned figure are the chromosomes 
    plus one global axes across all chromosomes
    """
    if plotfun is None:
        plotfun = plt.plot
    if grid is None:
        grid, parent_gridelement = get_chrom_grid(parent_gridelement=parent_gridelement)
    if fig is None:
        fig = plt.figure(1,figsize=(20,2))
    fig.subplots_adjust(wspace=0)
    #print color
    if color is None:
        print "changed to blue"
        color = mpl.rcParams['axes.color_cycle'][0]
    color = converter.to_rgb(color) 
    chroms = series.index.get_level_values(level=0).unique()
    for i,(chrom, gr) in enumerate(zip(chroms,grid)):
        if i == 0:
            ax = fig.add_subplot(gr)
        else:
            ax = fig.add_subplot(gr, sharey=ax)
            ax.set_yticks([])
        
        if i % 2 == 0:
            color1 = tuple(np.array(color)*0.6)
            if plot_xlabels:
                ax.set_xlabel(chrom.split("E")[1])
        else:
            color1 = color
        #print color1
        ax.set_frame_on(False)
        ax.set_xticks([])
        ax.set_color_cycle([color1])
    ylim = ax.get_ylim()
    #we could also get the parent element with gr.get_topmost_subplotspec()
    total_ax = fig.add_subplot(parent_gridelement)
    total_ax.set_frame_on(False)
    total_ax.spines['top'].set_visible(True)
    total_ax.set_ylim(ylim)
    total_ax.set_xticks([])
    if not ylabel and series.name is not None:
        ylabel = series.name
    total_ax.set_ylabel(ylabel,fontsize = 14)
    if plot_xlabels:
        total_ax.set_xlabel("Chromosomes",labelpad=25,fontsize = 14)
    return fig.get_axes()

converter =  mpl.colors.ColorConverter()
def plot_chrom_series(series,plotfun=None,grid=None,parent_gridelement=None,fig=None,
                      ylabel = "", plot_xlabels=True, color=None,title=None,rightlabel=None,**kwa):
    """
    axes of the returned figure are the chromosomes 
    plus one global axes across all chromosomes
    """
    if plotfun is None:
        plotfun = plt.plot
    if grid is None:
        grid, parent_gridelement = get_chrom_grid(parent_gridelement=parent_gridelement)
    if fig is None:
        fig = plt.figure(1,figsize=(20,2))
    fig.subplots_adjust(wspace=0)
    #print color
    if color is None:
        color = mpl.rcParams['axes.color_cycle'][0]
    color = converter.to_rgb(color) 
    chroms = series.index.get_level_values(level=0).unique()
    for i,(chrom, gr) in enumerate(zip(chroms,grid)):
        if i == 0:
            ax = fig.add_subplot(gr)
        else:
            ax = fig.add_subplot(gr, sharey=ax)
            ax.set_yticks([])
        
        if i % 2 == 0:
            color1 = tuple(np.array(color)*0.6)
            if plot_xlabels:
                ax.set_xlabel(chrom.split("E")[1])
        else:
            color1 = color
        #print color1
        ax.set_frame_on(False)
        ax.set_xticks([])
        ax.set_xlim([0,chrom_length[chrom]])
       # if chrom == 'CAE3':
            #print series.ix[chrom].index.values[~np.isnan(series.ix[chrom].values)]
            
        plotfun(series.ix[chrom].index.values,series.ix[chrom].values,color=color1,rasterized=True,**kwa)
    ylim = ax.get_ylim()
    #we could also get the parent element with gr.get_topmost_subplotspec()
    total_ax = fig.add_subplot(parent_gridelement)
    total_ax.set_frame_on(False)
    total_ax.spines['top'].set_visible(True)
    total_ax.set_ylim(ylim)
    total_ax.set_xticks([])
    if not ylabel and series.name is not None:
        ylabel = series.name
    total_ax.set_ylabel(ylabel,fontsize = 14)
    
    if plot_xlabels:
        total_ax.set_xlabel("Chromosomes",labelpad=25,fontsize = 14)
        
    if rightlabel is not None:
        ax2 = ax.twinx()
        ax2.set_ylabel(rightlabel, color=color1,fontsize=16)
        ax2.set_yticks([])
        ax2.spines['top'].set_visible(False)
        ax2.spines['right'].set_visible(False)
        ax2.spines['bottom'].set_visible(False)
        ax2.spines['left'].set_visible(False)
        
    if title is not None:
        plt.title(title)#, position=(0.5,1.02),va='bottom'

    return fig, total_ax, grid
```

```python
%%px --local
def gene_likelihood_chrom(chrom,mode="max"):
    """
    get max likelihood within each gene for all populations
    """
    result_dic = {}
    for i,pop0 in enumerate(pops2):
        for pop1 in [p for p in pops2 if p!=pop0]:
            xpclr_df = get_xpclr_chrom(chrom,pop0,pop1)
            result_dic.update({(pop0,pop1):gene_likelihood(xpclr_df,mode=mode)})
    return result_dic

def gene_likelihood(outlier_locations,mode="max"):
    """
    get genes that overlap with the multi-index of the provided data frame
    """
    modes = ["max","mean"]
    assert mode in modes
    gene_hit_df = pd.DataFrame()
    for chrom in outlier_locations.index.droplevel(1).unique():
        pos_rel_to_start = gene_df.ix[chrom].index.searchsorted(outlier_locations.ix[chrom].index).values
        pos_rel_to_end = np.searchsorted(gene_df.ix[chrom]["end"].values,outlier_locations.ix[chrom].index.values)
        in_gene = (pos_rel_to_start - pos_rel_to_end) == 1
        gene_hits = gene_df.ix[chrom].iloc[pos_rel_to_end[in_gene]]
        gene_hits['chrom'] = chrom
        gene_hits.set_index('chrom', append=True, inplace=True)
        gene_hits = gene_hits.reorder_levels(['chrom', 'pos'])
        gene_hits["score"] = outlier_locations[in_gene]["score"].values
        gene_hit_df = gene_hit_df.append(gene_hits)
    if mode == "max":    
        per_gene_df = gene_hit_df.groupby(lambda x:x).max()
        per_gene_df["length"] = per_gene_df["start"] - per_gene_df["end"]
        
    elif mode == "mean":
        per_gene_df = gene_hit_df.groupby(lambda x:x).mean()
        per_gene_df["length"] = per_gene_df["start"] - per_gene_df["end"]
    return per_gene_df
```

```python
def get_hiv_df(method_fn, poptuples):
    all_genes_df = pd.DataFrame(index=['all_genes'])
    pval_df = pd.DataFrame(index=hiv_cand)
    rand_k_df = pd.DataFrame(index=["k{}_".format(i) + s for s in hiv_cand for i in range(5)])
    rand_df = pd.DataFrame(index=["r{}_".format(i) + s for s in hiv_cand for i in range(5)])
    for pop0, pop1 in poptuples:
        s = pd.read_csv(jn(xpclr_ana_dir,method_fn.format(pop0,pop1)), sep='\t', usecols=[0,4], index_col=0, squeeze=True)
        s.name = "_".join([pop0, pop1])
        #all_genes_df =  pd.concat([all_genes_df,s.ix[all_genes_df.index]],axis=1)
        pval_df = pd.concat([pval_df,s.ix[pval_df.index]],axis=1)
        try:
            rand_k_df = pd.concat([rand_k_df,s.ix[rand_k_df.index]],axis=1)
            rand_df = pd.concat([rand_df,s.ix[rand_df.index]],axis=1)
        except:
            rand_k_df = None
            rand_df = None
    return pval_df, rand_k_df, rand_df

def benjamini_hochberg(col):
    return col.sort(inplace=False)*len(col)/np.arange(1,len(col)+1)

def above_benjamini_hochberg(col, thresh=0.05):
    bh_col = benjamini_hochberg(col)
    below_thrs = bh_col.values < thresh
    if np.sum(below_thrs) == len(col):
        #correct for the fact that argmin gives 0 if not found
        return bh_col
    else:
        return bh_col.iloc[:np.argmin(bh_col.values<thresh)]
```

```python
method, method_fn = 'mean_mean','xpclr_enrichment_{}_{}_mean_mean_exons_175000.pvals.tsv'
pval_df, rand_k_df, rand_df = get_hiv_df(method_fn, africa_poptuples)
```

```python
%%px --local
#general functions
def get_gen_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)

def get_subpop_df(gen_df,pop,poptype='pop3'):
    gen_df_pop = gen_df[mt[(mt[poptype]==pop)].index]
    #print len(gen_df_pop)
    #removing non-seggretating sites
    gen_df_pop = gen_df_pop[(gen_df_pop.mean(axis=1)>0)&(gen_df_pop.mean(axis=1)<2)]
    #print len(gen_df_pop)
    return gen_df_pop

def get_id_order(chrom):
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    with open(fn) as f:
        ids = f.readline().strip().split('\t')[2:]
    return ids


def get_ad_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_anc_der_including_fixed_{}.tsv".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)

def get_ad_seg_df(chrom):
    df = get_ad_df(chrom)
    return df[(df.sum(axis=1)>0)&\
                        (df.sum(axis=1)<2*df.notnull().sum(axis=1))]
```

```python
%%px --local

neff = json.load(open(jn(var_ana_dir,"n_eff_pop3.json")))
ballet_dir = eu("~/vervet_project/analyses/20150601_UG_ballet/_data")

def write_ballet_input(chrom):
    """
    Write BALLET input files according to
    http://www.personal.psu.edu/mxd60/Manual_BALLET_v1.0.pdf
    
    (M DeGiorgio, KE Lohmueller, R Nielsen (2014))
    """
    def get_poly(df):
        ac = df.sum(axis=1)
        n = 2*df.notnull().sum(axis=1)
        df = pd.DataFrame({'x':n.values-ac.values,'n':n.values}, index=ac.index.droplevel(0))
        df = df[['x','n']]
        df.index.name = 'position'
        df.dropna(inplace=True)
        return df[df['x']!=df['n']].astype(int)
    ad_df = get_ad_df(chrom)
    popgroup = ad_df.groupby(lambda id: mt['pop3'].ix[id],axis=1)
    for pop, df in popgroup:
        #start: redo pys:
        if pop != 'pys':
            continue
        df = df.iloc[:,:-1]
        #end: redo pys    
        poly_df = get_poly(df)
        
        poly_df.to_csv(jn(ballet_dir,"163_UG_ballet_polysub_{}_{}.tsv".format(pop,chrom)),sep='\t')
        
        pos_s = pd.Series(np.nan,index=poly_df.index)
        pos_s.name = 'centimorgan'
        map_values = map_s.ix[chrom].add(pos_s,fill_value=0).interpolate(method='values')
        
        #interpolating the first values
        first_idx, second_idx = map_s.ix[chrom].head(2).index.values
        first_value, second_value = map_s.ix[chrom].head(2).values
        map_values.ix[:first_idx] = (second_value - first_value) / (second_idx - first_idx) * \
                                                    map_values.ix[:first_idx].index + \
                                                    (first_value *second_idx - second_value * first_idx) / (second_idx - first_idx)
        #interpolating the last values
        secondlast_idx, last_idx = map_s.ix[chrom].tail(2).index.values
        secondlast_value, last_value = map_s.ix[chrom].tail(2).values
        map_values.ix[last_idx:] = (last_value - secondlast_value) / (last_idx - secondlast_idx) * \
                                                    map_values.ix[last_idx:].index + \
                                                    (secondlast_value *last_idx - last_value * secondlast_idx) / (last_idx - secondlast_idx)
        rho_s = map_values.ix[pos_s.index].diff().fillna(0)*(4*neff[pop])/100.
        rho_s.name = 'rate'
        
        rho_s.to_csv(jn(ballet_dir,"163_UG_ballet_rec_{}_{}.tsv".format(pop,chrom)),sep='\t',header=True)
```

```python
def get_hiv_df(method_fn, pops):
    all_genes_df = pd.DataFrame(index=['all_genes'])
    pval_df = pd.DataFrame(index=hiv_cand)
    rand_k_df = pd.DataFrame(index=["k{}_".format(i) + s for s in hiv_cand for i in range(5)])
    rand_df = pd.DataFrame(index=["r{}_".format(i) + s for s in hiv_cand for i in range(5)])
    for pop in pops:
        s = pd.read_csv(jn(ballet_dir,method_fn.format(pop)), sep='\t', usecols=[0,4], index_col=0, squeeze=True)
        s.name = pop
        #all_genes_df =  pd.concat([all_genes_df,s.ix[all_genes_df.index]],axis=1)
        pval_df = pd.concat([pval_df,s.ix[pval_df.index]],axis=1)
        try:
            rand_k_df = pd.concat([rand_k_df,s.ix[rand_k_df.index]],axis=1)
            rand_df = pd.concat([rand_df,s.ix[rand_df.index]],axis=1)
        except:
            rand_k_df = None
            rand_df = None
    return pval_df, rand_k_df, rand_df

def benjamini_hochberg(col):
    return col.sort(inplace=False)*len(col)/np.arange(1,len(col)+1)

def above_benjamini_hochberg(col, thresh=0.05):
    bh_col = benjamini_hochberg(col)
    below_thrs = bh_col.values < thresh
    if np.sum(below_thrs) == len(col):
        #correct for the fact that argmin gives 0 if not found
        return bh_col
    else:
        return bh_col.iloc[:np.argmin(bh_col.values<thresh)]
```
