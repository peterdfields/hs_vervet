```python
chromosomes = ["CAE" + str(i) for i in range(1,30)+["X","Y"]]
pops = ["aet","cyn","pyn","pys","sab","tan"]
colors = ["blue","magenta","cyan","green","orange","red"]
```

```python
chrom = "CAE29"
chroms = chromosomes[1:]
n_subplot = len(chroms)
fig=figure(1, figsize=(20,3*n_subplot))
hspace = 0.4
fig.subplots_adjust(hspace=hspace)
for i,chrom in enumerate(chroms):
    ax = fig.add_subplot(n_subplot,1,i+1)
    for pop,color in zip(pops,colors):
        res = pd.read_csv(eu("~/vervet_project/analyses/20140611_163_subpop_sweepfinder/"
                             "_data/SweeD_Report.163_ref3500_{}_{}".format(chrom,pop)),
                                                          sep="\t",skiprows=2,index_col=0)
        res["Likelihood"].plot(style=".",markersize=5,color=color,title=chrom)
```
