```python
import pandas as pd
import os,sys
sys.path.insert(0, os.path.expanduser('~/lib/python'))
sys.path.insert(0, os.path.join(os.path.expanduser('~/script')))
```

```python
def ancestral_stats(chromosome,method=233):
    def two_nucleotides(df,key):
        return df[(df[key]==df['ref'])+(df[key]==df['alt'])]
    snps=pd.DataFrame()
    for outgroup in ['macaque','human']:
        fname=os.path.expanduser("~/vervet_data/metadata/genotype_method_{0}/align_outgroup/vervet_snps_{1}_{2}.tsv".format(method,outgroup,chromosome))
        snps=snps.combine_first(pd.read_csv(fname,sep="\t",index_col=0))
    #ratio of snps, for which ancestral state can be derived only using macaque:
    ancestral_vervet=((snps['mac']==snps['alt'])+(snps['mac']==snps['ref'])).mean()
    
    #macaque and human agree
    notnull=snps[snps['mac'].notnull()*snps['human'].notnull()]
    agree=(notnull['mac']==notnull['human'])
    two_nucleotides1=two_nucleotides(notnull[agree],'mac')
    ancestral_vervet_and_human=len(two_nucleotides1)*1./len(snps)
    
    
    #macaque if it exists otherwise use human:
    mac_snps=snps[snps['mac'].notnull()]
    human_snps=snps[snps['mac'].isnull()*snps['human'].notnull()]
    mac_two=two_nucleotides(mac_snps,'mac')
    human_two=two_nucleotides(human_snps,'human')
    ancestral_vervet_or_human=(len(mac_two)+len(human_two))*1./len(snps)
    
    return (ancestral_vervet,ancestral_vervet_and_human,ancestral_vervet_or_human)
```

```python
chrom_ls=range(1,30)+['X','Y']
ancestral_vervet_ls=[]
ancestral_vervet_and_human_ls=[]
ancestral_vervet_or_human_ls=[]
for chrom in chrom_ls:
    (a,b,c)=ancestral_stats('CAE'+str(chrom),method=233)
    ancestral_vervet_ls.append(a)
    ancestral_vervet_and_human_ls.append(b)
    ancestral_vervet_or_human_ls.append(c)
```
