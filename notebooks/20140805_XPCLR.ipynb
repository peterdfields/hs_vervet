## prepare XPCLR input files

```python
%%px --local
def get_gen_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"UGr_163_ref3500_ref_alt_{}_2.012".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)
def get_allele_df(chrom):
    allele_df = pd.read_csv(jn(var_ana_dir,"GATK_UG_163_ref3500_{}_filtered_ancestral_allele_2.tsv".format(chrom)),
                            sep="\t",
                            index_col=[0,1],na_values=["NA"])
    return allele_df
def get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac):
    """
    get data-frames for two sub-populations were SNPs not seggregating in at least one of the two are removed
    """
    gen_df_pop01 = gen_df[meta_df[(meta_df["pop2"]==pop0)].index.union(meta_df[(meta_df["pop2"]==pop1)].index)]
    #print len(gen_df_pop)
    #removing non-seggretating sites
    gen_df_pop01 = gen_df_pop01[((gen_df_pop01.mean(axis=1)>0)&(gen_df_pop01.mean(axis=1)<2))]
    gen_df_pop0 = gen_df_pop01[meta_df[(meta_df["pop2"]==pop0)].index]
    gen_df_pop1 = gen_df_pop01[meta_df[(meta_df["pop2"]==pop1)].index]
    not_excess_missing = (gen_df_pop0.isnull().mean(axis=1)<max_missing_frac)&\
                            (gen_df_pop1.isnull().mean(axis=1)<max_missing_frac)
    
    #print len(gen_df_pop)
    return (gen_df_pop0[not_excess_missing],gen_df_pop1[not_excess_missing])

def gen_df_to_019(gen_df):
    df_0 = (gen_df > 0) + gen_df*0 #last term propagates nan
    df_0 = df_0.fillna(9)
    df_1 = (gen_df > 1) + gen_df*0
    df_1 = df_1.fillna(9)
    #to shuffle the haplotypes randomly
    shuffle_mat = np.random.randint(0,2,df_0.shape)
    df_0s = df_0 * shuffle_mat + df_1 * (1-shuffle_mat)
    df_1s = df_0 * (1-shuffle_mat) + df_1 * shuffle_mat
    assert (df_0s + df_1s == df_0 + df_1).all().all(), "the shuffling goes wrong"
    df_0s.columns = pd.MultiIndex.from_tuples(zip(df_0s.columns,[0]*len(df_0s.columns)))
    df_1s.columns = pd.MultiIndex.from_tuples(zip(df_1s.columns,[1]*len(df_1s.columns)))
    df01 = df_0s.join(df_1s).sort(axis=1).astype(np.int8)
    return df01
def save_geno_file(df_019,fn_base):
    #write header file
    with open(jn(xpclr_ana_dir,fn_base+".individuals"),"w") as f:
        f.write("\t".join([i for i in df_019.columns.droplevel(1).values])+"\n")
    df_019.to_csv(jn(xpclr_ana_dir,fn_base+".geno"),sep=" ",index=False,header=False)
    
def make_snp_file(df_019,allele_df,gen_map_df):
    snp_df = pd.DataFrame({"name":[i[0]+"_"+str(i[1]) for i in df_019.index.values]},index=df_019.index)
    snp_df["chrom"] = map(lambda s: s.split("E")[1],df_019.index.droplevel(1))
    snp_df["morgan"] = gen_map_df.ix[df_019.index]["centimorgan"]/100.
    snp_df["bp"] = df_019.index.droplevel(0)
    snp_df[["ref","alt"]] = allele_df.ix[df_019.index][["REF","ALT"]]
    return snp_df

def save_snp_file(snp_df,fn_base):
    snp_df.to_csv(jn(jn(xpclr_ana_dir,fn_base+".snp")),sep=" ",index=False,header=False)
    
def create_xpclr_input_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac):
    dfs = get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac)
    pops = [pop0, pop1]
    for i,j in [(1,0),(0,1)]:
        df_019 = gen_df_to_019(dfs[i])
        fn_base = "163_xpclr_{}_vs_{}_{}".format(pops[i],pops[j],chrom)
        save_geno_file(df_019,fn_base)
    snp_file0 = make_snp_file(df_019,allele_df,genetic_map)
    save_snp_file(snp_file0,fn_base)

def create_snp_file_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac):
    #only creates the snp file
    dfs = get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac)
    snp_file0 = make_snp_file(dfs,allele_df,genetic_map)
    save_snp_file(snp_file0,fn_base)    
def create_snp_files(chrom,max_missing_frac):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    for i,pop0 in enumerate(pops):
        for pop1 in pops[i+1:]:
            create_snp_file_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac)
    
def create_xpclr_input(chrom):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    for i,pop0 in enumerate(pops):
        for pop1 in pops[i+1:]:
            create_xpclr_input_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map)
            
def write_slice_xpclr_chrom(chrom,slice_size,overlap,max_missing_frac):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    #out = []
    for i, pop0 in enumerate(pops2):
        for pop1 in pops2[i+1:]: 
            df0, df1 = get_2_subpop_dfs(gen_df,pop0,pop1,max_missing_frac=max_missing_frac)
            df0_019 = gen_df_to_019(df0)
            df1_019 = gen_df_to_019(df1)
            snp_df = make_snp_file(df0_019,allele_df,genetic_map)
            #out.append(
            write_xpclr_in_slices(chrom,pop0,pop1,df0_019,df1_019,snp_df,slice_size,overlap)
            #)
    #return out

def write_xpclr_whole_chrom(chrom,max_missing_frac):
    gen_df = get_gen_df(chrom)
    allele_df = get_allele_df(chrom)
    genetic_map = load_genetic_map(chrom)
    #out = []
    for i, pop0 in enumerate(pops2):
        for pop1 in pops2[i+1:]: 
            create_xpclr_input_pop01(gen_df,chrom,pop0,pop1,allele_df,genetic_map,max_missing_frac)   
    
import json
def write_xpclr_in_slices(chrom,pop0,pop1,df0_019,df1_019,snp_df,slice_size,overlap):
    extensions = ["geno","geno","snp"]
    pops = [(pop0,pop1),(pop1,pop0),(pop0,pop1)]
    for j,df in enumerate([df0_019,df1_019,snp_df]):
        if j<2:
            continue
        fn_base = "163_xpclr_{}_vs_{}_{}".format(pops[j][0],pops[j][1],chrom)
        if j<2:
            with open(jn(xpclr_ana_dir,fn_base+".individuals"),"w") as f:
                f.write("\t".join([str(i) for i in df.columns.droplevel(1).values])+"\n")
        for i,start in enumerate(range(df.head(1).index.droplevel(0)[0],
                           df.tail(1).index.droplevel(0)[0],
                           slice_size-overlap)):
            end = min(start + slice_size, df.ix[chrom].index[-1])
            #try:
            df.ix[chrom].ix[start:end].to_csv(jn(xpclr_ana_dir,fn_base+"_{}".format(i)+"."+extensions[j]),
                                               sep=" ",header=False,index=False,float_format="%.7g")
            #except:
            #    return (df, chrom, start, end)
    #info_dic = {"slice_size":slice_size,"overlap":overlap,"n_slice":i+1,"info":"Input files for XPCLR with given slice size and overlap."}
    #json.dump(info_dic,open(jn(xpclr_ana_dir,fn_base+".INFO.json"),"w"))           
```

```python
def get_xpclr(chrom,pop0,pop1,i):
    xpclr_df = pd.read_csv(jn(xpclr_ana_dir,"163_xpclr_{}_vs_{}_{}_{}.result.xpclr.txt".format(pop0,pop1,chrom,i)),
                       header=None,names=["chrom","grid","n_snp","pos","morgan","score","max_s"],sep=" ")
    
    xpclr_df["chrom"] = chrom
    xpclr_df.set_index(["chrom","pos"],inplace=True)
    return xpclr_df

def get_xpclr_chrom(chrom,pop0,pop1):
    n_slice = get_info(chrom,pop0,pop1)["n_slice"]
    df = pd.concat([get_xpclr(chrom,pop0,pop1,i) for i in range(n_slice)])
    return df

def get_info(chrom,pop0,pop1):
    popsl = [pop0,pop1]
    popsl.sort()
    return json.load(open(jn(xpclr_ana_dir,"163_xpclr_{}_vs_{}_{}.INFO.json".format(popsl[0],popsl[1],chrom,i))))
```

```python
def plot_xpclr_chrom(xpclr_df,info,ax=None):  
    if ax is None:
        ax = plt.gca()
    #plt.set_gca(ax)
    plt.plot(xpclr_df.ix[chrom].index,xpclr_df["score"],".",markersize=5)
    for start in np.arange(xpclr_df.ix[chrom].index[0],xpclr_df.ix[chrom].index[-1],info["slice_size"]):
        plt.axvline(start,color="red")
    for end in np.arange(xpclr_df.ix[chrom].index[0]+info["slice_size"]-info["overlap"],xpclr_df.ix[chrom].index[-1],info["slice_size"]):
        plt.axvline(end,color="green")
```
