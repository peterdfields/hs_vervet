```python
%%px --local
def make_map_df(chrom):
    gen_df = get_gen_df(chrom)
    idx = gen_df.ix[chrom].index.union(map_s.ix[chrom].index)
    idx.name = "basepair"
    genetic_map = pd.DataFrame(index=idx)
    genetic_map["centimorgan"] = np.nan
    genetic_map.ix[map_s.ix[chrom].index,"centimorgan"] = map_s.ix[chrom].values
    genetic_map["163_pop_snp"] = 0
    genetic_map.ix[gen_df.ix[chrom].index,"163_pop_snp"] = 1
    genetic_map["vrc_snp"] = 0
    genetic_map.ix[map_s.ix[chrom].index,"vrc_snp"] = 1
    genetic_map["centimorgan"] = genetic_map["centimorgan"].interpolate(method='values')
    #interpolating the first values
    first_idx, second_idx = map_s.ix[chrom].head(2).index.values
    first_value, second_value = map_s.ix[chrom].head(2).values
    genetic_map.ix[:first_idx,"centimorgan"] = (second_value - first_value) / (second_idx - first_idx) * \
                                                genetic_map.ix[:first_idx,"centimorgan"].index + \
                                                (first_value *second_idx - second_value * first_idx) / (second_idx - first_idx)
    #interpolating the last values
    secondlast_idx, last_idx = map_s.ix[chrom].tail(2).index.values
    secondlast_value, last_value = map_s.ix[chrom].tail(2).values
    genetic_map.ix[last_idx:,"centimorgan"] = (last_value - secondlast_value) / (last_idx - secondlast_idx) * \
                                                genetic_map.ix[last_idx:,"centimorgan"].index + \
                                                (secondlast_value *last_idx - last_value * secondlast_idx) / (last_idx - secondlast_idx)
    genetic_map.index = pd.MultiIndex.from_tuples(zip([chrom]*len(idx),idx))
    genetic_map.index.names = ["chrom","pos"]
    return genetic_map
def save_genetic_map(gen_map_df):
    chrom = gen_map_df.index.levels[0][0]
    fn = jn(var_ana_dir,"UG_163_ref3500_genetic_map_{}.tsv".format(chrom))
    gen_map_df.index.names = ["chrom","pos"]
    gen_map_df.to_csv(fn, sep="\t",float_format="%.5g")
    
def save_genetic_map_chrom(chrom):
    map_df = make_map_df(chrom)
    save_genetic_map(map_df)
```

```python
%%px --local
#data functions
def get_ac(gen_df,mode="absorbing"):
    """
    mode = "sum" ... just sums up the entries, this means that nans are treated as 0
    mode = "mean" ... calculates average ignoring nans -> the spectrum is pulled towards
                        extreme frequencies
    mode = "absorbing" ... pull the spectrum to the left for frequencies <0.5 and to the
                            right for frequencies >0.5
    """
    n = gen_df.shape[1]
    if mode == "sum":
        return gen_df.sum(axis=1)
    elif mode == "mean":
        af = gen_df.mean(axis=1)/2.
        return (af*2*n).round()
    elif mode == "absorbing":
        mean_gt0p5 = (gen_df.mean(axis=1)/2.>0.5)
        na_val = 2.*mean_gt0p5
        gen_df[gen_df.isnull().mul(mean_gt0p5,axis=0)]=2
        gen_df[gen_df.isnull().mul(~mean_gt0p5,axis=0)]=0
        return gen_df.sum(axis=1)
    else:
        raise

def get_acs(gen_df,mode="absorbing"):
    """
    get allele count spectrum (histogram)
    """
    n = gen_df.shape[1]
    ac = get_ac(gen_df,mode=mode)
    bins = np.arange(-0.5,2*n+0.501)
    h, bins = np.histogram(ac,bins=bins)
    return h, bins
    
def acs_from_chrom(chrom,anc_der=True):
    """
    allele count spectrum (histogram)
    for either ancestral-derived (anc_der=True)
    or ref-alt alleles
    """
    gen_df = get_gen_df(chrom)
    if anc_der:
        anc_df = get_anc_df(chrom)
        df = anc_der_df(gen_df,anc_df)
        del anc_df
    else:
        df = gen_df
    del gen_df
    return get_acs(df)
```

```python
#plot function
def plot_hist(h,bins,legend=True):
    
    width = 0.7 * (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    plt.bar(center, h, align='center', width=width,color="black")
    ax = plt.gca()
    ax.set_xlim(bins[0],bins[-1])
    
    #calculate and plot theta waterson
    n = int(bins[-1])
    theta_waterson = h.sum()*1./sum([1./i for i in range(1,n)])
    x = np.arange(2,2*n)
    label ="neutral spectrum\n w/ wattersons theta"
    line, = plt.plot(x,theta_waterson/x,linewidth=2.5, linestyle="--",color=(0.8,0.8,0),
                label=label)
    if legend:
        ax.legend()
    return line,label
```

#### Pairwise differences

```python
%%px --local
def pairwise_diff_numpy(gen_arr):
    """Squared pairwise distances between all 
    columns of 0,1,2 genotype array arr.
    This matrix based function is at least 10 
    times faster than iterating over columns.
    """
    gen_arr = gen_arr.astype(np.float64)-1
    #compare heterozygous with hom alt
    mat1 = np.where(gen_arr==0,-1,gen_arr)
    mat1[np.isnan(gen_arr)]=0
    #and hom ref
    mat2 = np.where(gen_arr==0,1,gen_arr)
    mat2[np.isnan(gen_arr)]=0
    #account for heterozygous comparisons
    mat3 = np.where(gen_arr==0,1,0)
    #don't count nan comparisons
    n = np.dot((~np.isnan(gen_arr)*1.).T,~np.isnan(gen_arr)*1.)
    B = (np.dot(mat1.T,mat1)+np.dot(mat2.T,mat2))/2.
    het = np.dot(mat3.T,mat3)
    return ((n- B)+het)/2.

def pairwise_diff_mat(df):
    """
    Calculate pairwise difference data frame.
    For Genotype 0,1,2 data frame.
    Uses numpy matrix multiplication.
    """
    diff = pairwise_diff_numpy(df.values)
    return pd.DataFrame(diff,index=df.columns,columns=df.columns)
```

```python
%%px --local
def get_acs2D(gen_df1,gen_df2,mode="mean"):
    """
    get 2D allele count spectrum (histogram)
    for 2 populations
    """
    n1 = gen_df1.shape[1]
    n2 = gen_df2.shape[1]
    ac1 = get_ac(gen_df1,mode=mode)
    ac2 = get_ac(gen_df2,mode=mode)
    bins = [np.arange(-0.5,2*n1+0.501),np.arange(-0.5,2*n2+0.501)]
    h = np.histogram2d(ac1.values,ac2.values,bins=bins)
    return h
def get_pop_hists(chrom):
    gen_df = get_gen_df(chrom)
    anc_df = get_anc_df(chrom) 
    hists = {}
    for pop, p_meta_df in meta_df.groupby("population"):
        p_gen_df = gen_df.ix[:,p_meta_df.index.unique()]
        #remove fixed
        p_gen_df = anc_der_df(p_gen_df,anc_df)
        p_gen_df = p_gen_df[(p_gen_df.sum(axis=1)>0)*(p_gen_df.sum(axis=1)<2*p_gen_df.notnull().sum(axis=1))]
        hists.update({pop:get_acs(p_gen_df,mode="mean")})
    return hists
def get_pop_hists_1d2d(chrom):
    hists = {}
    hists2d = {}
    gen_df = get_gen_df(chrom)
    anc_df = get_anc_df(chrom) 
    gen_df = anc_der_df(gen_df,anc_df)
    for i,pop in enumerate(pops):
        p_gen_df = gen_df.ix[:,meta_df[meta_df["population"]==pop].index.unique()]
        p_gen_df_non_fixed = p_gen_df[(p_gen_df.sum(axis=1)>0)&\
                                        (p_gen_df.sum(axis=1)<2*p_gen_df.notnull().sum(axis=1))]
        hists.update({pop:get_acs(p_gen_df_non_fixed,mode="mean")})
        for pop2 in pops[i+1:]:
            p2_gen_df = gen_df.ix[:,meta_df[meta_df["population"]==pop2].index.unique()]
            acs2D = get_acs2D(p_gen_df,p2_gen_df)
            hists2d.update({(pop,pop2):acs2D})
    return hists,hists2d
def get_pop2_hists_1d2d(chrom):
    """
    here Caribbean sabaeus are treated as seperate population
    """
    hists = {}
    hists2d = {}
    gen_df = get_gen_df(chrom)
    anc_df = get_anc_df(chrom) 
    gen_df = anc_der_df(gen_df,anc_df)
    for i,pop in enumerate(pops2):
        p_gen_df = gen_df.ix[:,meta_df[meta_df["pop2"]==pop].index.unique()]
        p_gen_df_non_fixed = p_gen_df[(p_gen_df.sum(axis=1)>0)&\
                                        (p_gen_df.sum(axis=1)<2*p_gen_df.notnull().sum(axis=1))]
        hists.update({pop:get_acs(p_gen_df_non_fixed,mode="mean")})
        for pop2 in pops2[i+1:]:
            p2_gen_df = gen_df.ix[:,meta_df[meta_df["pop2"]==pop2].index.unique()]
            acs2D = get_acs2D(p_gen_df,p2_gen_df)
            hists2d.update({(pop,pop2):acs2D})
    return hists,hists2d
```

```python
%%px --local
def fixed_diff_genes(chrom,ac_threshold=1):
    """
    get dictionary which contains for each population pair
    the genes in which carry snps with fixed differences (nearly all genes)
    along with the density of such snps
    """
    gen_df = get_gen_df(chrom)
    fixed_diff_gene_dic = {}
    for i,pop1 in enumerate(pops2):
        gen_df1 = gen_df.ix[:,meta_df[meta_df["pop2"]==pop1].index.unique()]
        for pop2 in pops2[i+1:]:
            gen_df2 = gen_df.ix[:,meta_df[meta_df["pop2"]==pop2].index.unique()]
            fixed_diff = get_fixed_diff(gen_df1,gen_df2,ac_threshold=ac_threshold)
            fixed_diff_gene_dic.update({pop1+"_"+pop2:get_gene_hits(fixed_diff)})
    return fixed_diff_gene_dic
def get_fixed_diff(gen_df1,gen_df2,ac_threshold=1):
    """
    get SNPs for which the two populations have fixed differences
    ac_threshold gives the threshold for what is considered a fixed difference
    ac_threshold=1 means that difference does not need to be totally fixed
    but the allele count can be up to 1 in one population and must be more than n-1 in the other population
    """
    n1 = gen_df1.shape[1]
    n2 = gen_df2.shape[1]
    ac1 = get_ac(gen_df1,mode="mean")
    ac2 = get_ac(gen_df2,mode="mean")
    fixed_diff = pd.Series(True,index=gen_df1[(ac1<=ac_threshold)&\
                                             (ac2>=n2-ac_threshold)|\
                                             (ac1>=n1-ac_threshold)&(ac2<=ac_threshold)].index)
    return fixed_diff
def get_gene_hits(outlier_locations):
    """
    get genes that overlap with the multi-index of the provided data frame
    """
    gene_hit_df = pd.DataFrame()
    for chrom in outlier_locations.index.droplevel(1).unique():
        pos_rel_to_start = gene_df.ix[chrom].index.searchsorted(outlier_locations.ix[chrom].index).values
        pos_rel_to_end = np.searchsorted(gene_df.ix[chrom]["end"].values,outlier_locations.ix[chrom].index.values)
        in_gene = (pos_rel_to_start - pos_rel_to_end) == 1
        gene_hits = gene_df.ix[chrom].iloc[pos_rel_to_end[in_gene]]
        gene_hits['chrom'] = chrom
        gene_hits.set_index('chrom', append=True, inplace=True)
        gene_hits = gene_hits.reorder_levels(['chrom', 'pos'])
        gene_hit_df = gene_hit_df.append(gene_hits)
    n_hits = gene_hit_df.groupby(lambda x:x).apply(len)
    gene_hit_df = gene_hit_df.drop_duplicates()
    gene_hit_df["length"] =(gene_hit_df["end"]-gene_hit_df.index.droplevel(0).values)
    gene_hit_df["n_hits"] = n_hits
    gene_hit_df["hits_per_kb"] = n_hits *1000. / gene_hit_df["length"]
    return gene_hit_df
```

```python
#%%px --local
def get_fst(chrom,pop0,pop1):
    fn = eu("~/vervet_project/analyses/20140521_popgen_stats/"
            "_data/163_fst_{}_{}_{}.weir.fst".format(pop0,pop1,chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",squeeze=True,na_values=["nan","-nan"])
    return df

def mean_fst_df(chrom):
    poptuples = [[(pop0,pop1) for pop1 in pops[i+1:]] for i,pop0 in enumerate(pops)]
    poptuples = [item for sublist in poptuples for item in sublist]
    index = pd.MultiIndex.from_tuples(poptuples, names=['pop0', 'pop1'])
    out_df = pd.DataFrame(columns=[str(f) for f in np.arange(0.1,0.501,0.1)],index=index)
    n_snp_df = pd.DataFrame(columns=[str(f) for f in np.arange(0.1,0.501,0.1)],index=index)
    gen_df = get_gen_df(chrom)
    for pop01 in poptuples:
        pop0 = pop01[0]
        pop1 = pop01[1]
        fst_df = get_fst(chrom,pop0,pop1)
        gen_df_2pop = gen_df[meta_df[(meta_df["population"]==pop0)|(meta_df["population"]==pop1)].index]
        af = gen_df_2pop.mean(axis=1)/2.
        maf = (af<0.5)*af + (af>=0.5)*(1-af)
        #remove fixed snps
        fst_df = fst_df.ix[maf[maf>0].index.intersection(fst_df.index)]
        fst_means = []
        n_snps = []
        for val, df in fst_df.groupby(((maf+0.049)*10).round()):
            maf_slice = np.array([val-1,val])/10.
            fst_means.append(df.mean())
            n_snps.append(len(df))
        out_df.ix[(pop0,pop1)] = fst_means
        n_snp_df.ix[(pop0,pop1)] = n_snps
    return out_df,n_snp_df
```

```python
def mul(df1,df2):
    assert list(df1.index) == list(df2.index)
    assert list(df1.columns) == list(df2.columns)
    return pd.DataFrame(df1.values*df2.values,index=df1.index,columns=df1.columns)
def reduce_mean_fst(mean_fsts):

    numerator = reduce(lambda x,y: x+y,[mul(el[0],el[1]) for el in mean_fsts])
    denominator = reduce(lambda x,y:x+y, [el[1] for el in mean_fsts])
    return numerator.div(denominator)
```

```python
#%%px --local
def get_subpop_df(gen_df,pop):
    gen_df_pop = gen_df[meta_df[(meta_df["population"]==pop)].index]
    #print len(gen_df_pop)
    #removing non-seggretating sites
    gen_df_pop = gen_df_pop[(gen_df_pop.mean(axis=1)>0)&(gen_df_pop.mean(axis=1)<2)]
    #print len(gen_df_pop)
    return gen_df_pop

def calc_r2(gen_df_pop,snp_distance,tolerance=0.1,rand_deviation=0):
    """
    calculate correlation coefficient between snps
    
    tolerance ... fraction of deviation in distance 
                    that is tolerated (0.1 means +-10%)
    rand_deviation ... add a random deviation so that 
                        not too many neighboring snps have 
                                            the same partner
    """
    chrom = gen_df_pop.index[0][0]
    #snps that have a counter part snp_distance bp further
    last_idx = gen_df_pop.ix[chrom].index.searchsorted(gen_df_pop.ix[chrom].index[-1]-snp_distance)
    idx = gen_df_pop.ix[chrom].index[:last_idx-1]
    # 

    dev_vec = np.random.randint(-1-rand_deviation,1+rand_deviation,len(idx))
    #index snp_distance base-pairs apart
    second_idx = gen_df_pop.ix[chrom].index[idx.searchsorted(idx+snp_distance)+dev_vec]
    real_dist = second_idx.values - idx.values
    #print "total pairs",len(real_dist)
    useable = (real_dist > snp_distance*(1-tolerance))&(real_dist < snp_distance*(1+tolerance))
    idx = idx[useable]
    second_idx = second_idx[useable]
    real_dist = real_dist[useable]
    #print "pairs within tolerance",len(real_dist)
    df1 = gen_df_pop.ix[chrom].ix[idx]
    df2 = gen_df_pop.ix[chrom].ix[second_idx]
    index = df1.index + real_dist/2
    df2.index = index
    df1.index = index + 0
    not_null = df1.notnull().mul(df2.notnull())
    max_missing_frac = 0.4
    not_null = not_null.mul((not_null.mean(axis=1)>(1-max_missing_frac)),axis=0)
    df1 = df1[not_null]
    df2 = df2[not_null]

    covariance = pd.DataFrame(df1.values*df2.values, columns=df1.columns, index=df1.index).mean(axis=1) -\
                        df1.mean(axis=1) *\
                        df2.mean(axis=1)
    def var(df):
        return (df**2).mean(axis=1)- df.mean(axis=1)**2
    
    r2 =  pd.DataFrame({"r2":covariance**2/(var(df1)*var(df2)),"dist":real_dist})
    try:
        r2.index = pd.MultiIndex.from_tuples(zip([chrom]*len(r2.index),r2.index))
    except TypeError:
        print chrom,snp_distance,r2
    return r2
```

```python
# better calculate ld for random snp pairs
def calc_mean_r2(chrom):
    distances = [100,200,300,400,500,750,1000,1500,2000,3000,4000,
                 5000,7500,10000,15000,25000,50000,75000,100000,250000,500000]
    mean_r2_df = pd.DataFrame(columns=pops,index=distances)
    n_pairs_df = pd.DataFrame(columns=pops,index=distances)
    gen_df = get_gen_df(chrom)
    for pop in pops:
        gen_df_pop = get_subpop_df(gen_df,pop)
        for distance in distances:
            r2 = calc_r2(gen_df_pop,snp_distance=distance,tolerance=0.1,rand_deviation=0)
            mean_r2_df.ix[distance,pop] = r2["r2"].mean()
            n_pairs_df.ix[distance,pop] = len(r2)
    return mean_r2_df,n_pairs_df
```

##### Get the empirical value of r2 for unlinked sites at hardy weinberg as a function of sample size and allele frequency spectrum

```python
def get_random_genotype_df(gen_df,sfs):
    """
    generate a df with the same index as the input df
    but with allele frequencies randomly drawn from the site frequency spectrum
    for each site genotypes are then drawn assuming hardy weinberg equilibrium
    thus, all snps are unlinked

    sfs: np.histogram result, tuple of counts and bins
    """
    n_sites = gen_df.shape[0]
    af_probas = sfs[0].astype(float)/sum(sfs[0])
    n_chroms = len(af_probas) - 1 
    af_values = (sfs[1][:-1]+0.5)/n_chroms
    af_vec = np.random.choice(af_values, 
                                p=af_probas, 
                                size=n_sites)
    n = n_chroms/2
    assert n == gen_df.shape[1]
    unlinked_gen_array = np.array(map(lambda af: get_random_genotypes(af,n), af_vec))
    return pd.DataFrame(unlinked_gen_array,columns=gen_df.columns,index=gen_df.index)

def get_random_genotypes(af,n):
    return np.random.choice([0,1,2], 
                            p=[(1-af)**2,2*af*(1-af),af**2], 
                            size=n)

#use empirical allele frequency probability
def get_r2_for_unlinked_sites(gen_df,pop):
    subpop_df = get_subpop_df(gen_df,pop)
    unlinked_subpop_df = get_random_genotype_df(subpop_df,hists1d[pop])
    snp_distance = 5000 #choice of this value should not matter
    r2 = calc_r2(unlinked_subpop_df,snp_distance,tolerance=0.1,rand_deviation=0)
    return r2.mean()["r2"]
```

```python
%%px --local
def d_statistic(chrom):
    """
    "normalised Fst deviation" defined in
    Akey et al. 2010
    """
    d = pd.DataFrame(index=[])
    for i, pop0 in enumerate(pops):
        for j, pop1 in enumerate(pops[:i]+pops[i+1:]):
            fst = get_fst(chrom,
                            (pop0 if i<=j else pop1),
                            pop1 if i<=j else pop0)
            if j == 0:
                d0 = pd.DataFrame(index=fst.index)
            fst.name = pop0+"_"+pop1
            d0 = d0.join((fst-fst.mean())/fst.var(),how='outer')
        #print "fraction of sites kept",len(d0[d0.notnull().sum(axis=1)>1])*1./len(d0)
        #only keep sites with at least 2 not-nan fst_comparisons
        d0 = d0[d0.notnull().sum(axis=1)>1]
        di = d0.sum(axis=1)
        di.name = pop0
        d[di.name] = di
    return d

def windowed_d(d,maf,min_maf=0.1,window_size=10**4,min_snp_per_window=5):
    """
    Calculate the mean of d statistic in  non-overlapping
    windows over the genome.
    """
    maf0 = maf[maf.index.intersection(d.index)] 
    window_d = d[maf>0.1].groupby(lambda x: x[1]//window_size).\
                        filter(lambda x: len(x) >= min_snp_per_window).\
                            groupby(lambda x: x[1]//window_size).mean()
    #centre index at window centre
    window_d.index = pd.MultiIndex.from_tuples(zip([d.index[0][0]]*len(window_d),
                                                    window_d.index * window_size + window_size/2.))
    return window_d

def calc_3window_d(chrom):
    windows = [10000,100000,1000000]
    d = d_statistic(chrom)
    af = read_af_tsv(chrom)
    maf = (af<0.5)*af + (af>=0.5)*(1-af)
    return tuple([windowed_d(d,maf,window_size=window) for window in windows])

def save_3window_d(window_d3s):
    windows = [10000,100000,1000000]
    for chrom,res in zip(chromosomes,window_d3s):
        for w,df in zip(windows,res):
            df.to_csv(jn(man_ana_dir,"akey_d_window_{}_chrom_{}".format(w,chrom)),sep="\t")

def reduce_3window_d(window_d3s):
    total_dfs = []
    for i in range(3):
        total_dfs.append(pd.concat([res[i] for res in window_d3s]))
    return tuple(total_dfs)

def get_quantiles(window_d3,quantiles=(0.9,0.95,0.99)):
    qt = []
    for df in window_d3:
        qt.append(df.ix[autosomes].quantile(quantiles))
    return tuple(qt)      
```

```python
def get_gene_feature(outlier_locations,mode="max"):
    """
    get genes that overlap with the multi-index of the provided data frame
    """
    modes = ["max","sum","mean"]
    assert mode in modes
    gene_hit_df = pd.DataFrame()
    for chrom in outlier_locations.index.droplevel(1).unique():
        pos_rel_to_start = gene_df.ix[chrom].index.searchsorted(outlier_locations.ix[chrom].index).values
        pos_rel_to_end = np.searchsorted(gene_df.ix[chrom]["end"].values,outlier_locations.ix[chrom].index.values)
        in_gene = (pos_rel_to_start - pos_rel_to_end) == 1
        gene_hits = gene_df.ix[chrom].iloc[pos_rel_to_end[in_gene]]
        gene_hits['chrom'] = chrom
        gene_hits.set_index('chrom', append=True, inplace=True,drop=False)
        gene_hits = gene_hits.reorder_levels(['chrom', 'pos'])
        

        for column in outlier_locations.columns:
            gene_hits[column] = outlier_locations.ix[chrom][in_gene][column].values

        gene_hit_df = gene_hit_df.append(gene_hits)
    group = gene_hit_df.groupby(["chrom","gene_id"])
    if mode == "max":    
        return group.max()
    elif mode == "sum":
        return group.sum()
    elif mode == "mean":
        return group.mean()
```

```python
%%px --local
def gene_likelihood_chrom(chrom):
    """
    get max likelihood within each gene for all populations
    """
    result_dic = {}
    for pop in pops:
        sweed_df = get_sweed_df(chrom,pop)
        result_dic.update({pop:gene_likelihood(sweed_df)})
    return result_dic

def get_sweed_df(chrom,pop):
    sweed_res = pd.read_csv(eu("~/vervet_project/analyses/20140611_163_subpop_sweepfinder/"
                             "_data/SweeD_Report.163_ref3500_{}_{}".format(chrom,pop)),
                                                          sep="\t",skiprows=2,index_col=0)
    sweed_res.index = pd.MultiIndex.from_tuples(zip([chrom]*len(sweed_res.index),sweed_res.index))
    return sweed_res

def gene_likelihood(outlier_locations,mode="max"):
    """
    get genes that overlap with the multi-index of the provided data frame
    """
    modes = ["max","sum"]
    assert mode in modes
    gene_hit_df = pd.DataFrame()
    for chrom in outlier_locations.index.droplevel(1).unique():
        pos_rel_to_start = gene_df.ix[chrom].index.searchsorted(outlier_locations.ix[chrom].index).values
        pos_rel_to_end = np.searchsorted(gene_df.ix[chrom]["end"].values,outlier_locations.ix[chrom].index.values)
        in_gene = (pos_rel_to_start - pos_rel_to_end) == 1
        gene_hits = gene_df.ix[chrom].iloc[pos_rel_to_end[in_gene]]
        gene_hits['chrom'] = chrom
        gene_hits.set_index('chrom', append=True, inplace=True)
        gene_hits = gene_hits.reorder_levels(['chrom', 'pos'])
        gene_hits["likelihood"] = outlier_locations[in_gene]["Likelihood"].values
        gene_hit_df = gene_hit_df.append(gene_hits)
    #if mode == "max":    
    return gene_hit_df.groupby(lambda x:x).max()
    #elif mode == "sum":
    #    return gene_hit_df.groupby(lambda x:x).sum()
```

##### 2) list of sites with highest likelihoods

```python
def get_top_hits(chrom,sweed_df,top_n,min_dist=10000):
    top_hits = pd.DataFrame()
    exclude = np.array([])
    for row in sweed_df.sort("Likelihood",ascending=False).ix[chrom].iterrows():
        #print np.absolute(exclude-row[0])>min_dist
        #print row
        if (np.absolute(exclude-row[0])>min_dist).all():
            exclude = np.append(exclude,row[0])
            top_hits = top_hits.append(pd.DataFrame(row[1]).T)
        if len(top_hits) == top_n:
            break
    top_hits.index = pd.MultiIndex.from_tuples(zip([chrom]*top_n,top_hits.index))
    top_hits.index.names = ["chrom", "pos"]
    
    return get_gene_info(top_hits)

def get_top_hits_all_chrom(pop,top_n,min_dist=10000):
    top_hits_ls = []
    for chrom in chromosomes:
        sweed_df = get_sweed_df(chrom,pop)
        top_hits_ls.append(get_top_hits(chrom,sweed_df,top_n,min_dist=min_dist))
    all_top_hits = pd.concat(top_hits_ls)
    return all_top_hits.sort("Likelihood",ascending=False).iloc[:top_n]

def get_gene_info(hit_df):
    """
    For each index location in hit_df add the location of the closest
    gene to the data frame.
    All entries in the df must be on the same chromosome.
    """
    
    hit_df = hit_df.copy()
    chrom_ls = hit_df.index.droplevel(1).unique()
    assert len(chrom_ls) == 1, "Only implemented for data from single chromosome, but len(chrom_ls)=" + str(len(chrom_ls))
    chrom = chrom_ls[0]
    loc_gene_df = gene_df.ix[chrom]
    loc_gene_df = loc_gene_df.append(pd.DataFrame(np.nan,index=[inf],columns=loc_gene_df.columns))
    pos_rel_to_start = loc_gene_df.index.searchsorted(hit_df.ix[chrom].index).values
    pos_rel_to_end = np.searchsorted(loc_gene_df["end"].values,hit_df.ix[chrom].index.values)
    in_gene = (pos_rel_to_start - pos_rel_to_end) == 1
    hit_df["in_gene_id"] = np.nan
    hit_df.ix[in_gene,"in_gene_id"] = loc_gene_df.iloc[pos_rel_to_end[in_gene]]["gene_id"].values
    hit_df["in_gene_len"] = np.nan
    hit_df.ix[in_gene,"in_gene_len"] = loc_gene_df.iloc[pos_rel_to_end[in_gene]]["end"].values - \
                                        loc_gene_df.iloc[pos_rel_to_end[in_gene]].index.values
    hit_df["in_gene_rel_pos"] = np.nan
    #print hit_df.ix[in_gene].ix[chrom].index.values - loc_gene_df.iloc[pos_rel_to_end[in_gene]].index.values
    hit_df.ix[in_gene,"in_gene_rel_pos"] = ( hit_df.ix[in_gene].ix[chrom].index.values - \
                                            loc_gene_df.iloc[pos_rel_to_end[in_gene]].index.values)/\
                                            (loc_gene_df.iloc[pos_rel_to_end[in_gene]]["end"].values - \
                                             loc_gene_df.iloc[pos_rel_to_end[in_gene]].index.values)
    
    hit_df["upstream_gene_id"] = loc_gene_df.iloc[pos_rel_to_end-1]["gene_id"].values
    hit_df["upstream_gene_end_dist"] = hit_df.ix[chrom].index.values  - loc_gene_df.iloc[pos_rel_to_end-1]["end"].values
    hit_df["upstream_gene_len"] = loc_gene_df.iloc[pos_rel_to_end-1]["end"].values - loc_gene_df.iloc[pos_rel_to_end-1].index.values
    #print pos_rel_to_start
    hit_df["downstream_gene_id"] = loc_gene_df.iloc[pos_rel_to_start]["gene_id"].values
    hit_df["downstream_gene_start_dist"] = loc_gene_df.iloc[pos_rel_to_start].index.values - hit_df.ix[chrom].index.values
    hit_df["downstream_gene_len"] = loc_gene_df.iloc[pos_rel_to_start]["end"].values - \
                                        loc_gene_df.iloc[pos_rel_to_start].index.values
    hit_df["closest_gene"] = np.nan
    hit_df.ix[in_gene,"closest_gene"] = hit_df.ix[in_gene,"in_gene_id"]
    hit_df.ix[~in_gene,"closest_gene"] = (hit_df.ix[~in_gene,"upstream_gene_end_dist"] < \
                                          hit_df.ix[~in_gene,"downstream_gene_start_dist"])*hit_df.ix[~in_gene,"upstream_gene_id"] +\
                                         (hit_df.ix[~in_gene,"upstream_gene_end_dist"] >= \
                                          hit_df.ix[~in_gene,"downstream_gene_start_dist"])*hit_df.ix[~in_gene,"downstream_gene_id"]
    return hit_df
```
