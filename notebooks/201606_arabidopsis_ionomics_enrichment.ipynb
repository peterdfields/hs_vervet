```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from enrichme import enrichme as em
from enrichme import pandas_util as pu
from pypopgen.plot import genomeplot as gp

eu = os.path.expanduser
jn = os.path.join
meta_dir = eu("~/1001genomes_project/Hannes/metadata")
```

```python
gene_df = pd.read_csv(eu("~/1001genomes_project/ARCHIVE/Hannes/data/annotation/at_annotation_minimal.tsv"),sep="\t",index_col=[0,1])
```

```python
#load data
```

```python
ion_df = pd.read_csv(eu("~/1001genomes_project/ARCHIVE/Hannes/data/gwas_results/LIMIX_BOX_COX_NO_PCA_ALL_INPUT_ENRICHMENT.csv"))
ion_df.set_index(["chrom","pos"],inplace=True)
```

```python
ions = list(ion_df.columns[2:])
```

```python
ion_df10 = ion_df[ion_df['MAF']>0.1]
```

```python
ion10_score_per_gene = pu.data_per_feature(ion_df10,gene_df, feature_name='gene_id')
```

```python
ion10_max_per_gene = ion10_score_per_gene.groupby('gene_id').max()
ion10_mean_per_gene = ion10_score_per_gene.groupby('gene_id').mean()
```

```python
ion10_max_per_gene.to_csv(eu("~/Akademisches/Projects/1001genomes/scratch/ion10_max_per_gene.tsv"), sep='\t')
```

```python
ion10_mean_per_gene.to_csv(eu("~/Akademisches/Projects/1001genomes/scratch/ion10_mean_per_gene.tsv"), sep='\t')
```

```python
for ion in ions:
    with open(eu("~/Akademisches/Projects/1001genomes/scratch/ion_{}_genes.txt".format(ion)),'w') as f:
        for v in ion10_mean_per_gene[ion].sort(inplace=False).index.values:
            f.write(v+'\n')
```
