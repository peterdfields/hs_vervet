## Joint alignment nucleus + siv sequence

```python
meta_fn = os.path.expanduser("~/vervet_project/metadata/163_population_unfiltered_read_mate_path_phred33.csv")
meta_df = pd.read_csv(meta_fn)
ucla_ids = list(meta_df["ucla_id"].unique())
ucla_ids.remove("VSAM0021")
```

```python
siv_cov = pd.read_csv(eu("~/vervet_project/analyses/20141017_align_mito_siv/_data/siv_coverage.tsv"),sep="\t",index_col=[0,1],names=ucla_ids)
```

```python
mean_cov_strains = (siv_cov*1./mt.ix[ucla_ids]["coverage"]).groupby(lambda ix: ix[0]).mean()#.shape#.div(,axis=0)
```

```python
loc_mt = mt.ix[ucla_ids]
```

```python
mean_cov = mean_cov_strains.ix["DJ048201"]
#plt.scatter(mean_cov.ix["DJ048201"],mt.ix[ucla_ids]["species"])
```

```python
countries = loc_mt.country.unique()
country_num = pd.Series(index=mean_cov.index) 
for i,country in enumerate(countries):
    country_num.ix[loc_mt["country"]==country] = i
```

```python
ax = plt.gca()
#plt.plot(country_num,mean_cov,".")
plt.scatter(country_num,mean_cov)
plt.xticks(range(len(countries)),countries, rotation='vertical')
ax.set_xlim([-1,len(countries)])
ax.set_ylim([0,0.005])
#ax.set_yscale('exp')
plt.title("normalised SIV coverage")
```
