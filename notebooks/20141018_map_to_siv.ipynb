```python
#normalised by average autosomal coverage
mean_cov = cov_df.groupby(lambda idx:idx[0]).mean().div(mt["coverage"],axis=1).T
del mean_cov["gi|107736076|ref|NC_008066.1|"]
```

```python
#all species have highest coverage with 
(mean_cov[mean_cov>0.2]).groupby(mt["species"]).mean()
```

```python
#mapping against all viral sequences is perfactly correlated
plt.scatter(mean_cov.ix[">DJ048201|DJ048201"],mean_cov.ix[">HQ378594|HQ378594"])
```

```python
# work with mean across all viral sequences
mean_cov = mean_cov.mean(axis=1)
```

```python
mt.country.value_counts()
```

```python
country_mean = mean_cov.groupby(mt.country).mean()
country_std = mean_cov.groupby(mt.country).std()
```

```python
ax = plt.gca()
ax.errorbar(range(len(country_mean)), country_mean, yerr=country_std, fmt='o')
plt.xticks(range(len(country_mean)),country_mean.index, rotation='vertical')
ax.set_xlim([-1,len(country_mean)])
```

```python
countries = mt.country.unique()
```

```python
country_num = pd.Series(index=mean_cov.index) 
for i,country in enumerate(countries):
    country_num.ix[mt["country"]==country] = i
```

```python
def get_siv_pheno(phen_dic):
    try:
        return phen_dic["SIV"]
    except KeyError:
        return -1
```

```python
siv_pheno = mt["phenotypes"].apply(get_siv_pheno)[mt["phenotypes"].apply(get_siv_pheno) != -1]
siv_pheno.name = "siv_pheno"
```

```python
mean_cov.name = "siv_cov"
cov_pheno = pd.concat([mean_cov.ix[siv_pheno.index], siv_pheno],axis=1).sort("siv_cov")
```

```python
cov_pheno["cov>0.2"] = cov_pheno["siv_cov"]>0.2
```

```python
cov_pheno["country"] = mt["country"]
```
