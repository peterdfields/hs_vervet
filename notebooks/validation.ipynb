```python
def get_snp_df(vcf_to_validate,chrom,start,end,ucla_id,omit_non_segregating=False,omit_indels=False):
    # functions
    def get_alleles(record,ucla_id):
        """
        returns (ref_allele,genotype_allele_1,genotype_allele_2)
        """
        #print 'VCF ref:', record.REF,'ref:' ,ref[chrom][int(record.POS)-1]
        pos = int(record.POS)
        snv_len=len(record.REF)
        gt=record.genotype(ucla_id)['GT']
        #if gt is None:
        #    print record.genotype(ucla_id)
        ref_allele = ref[chrom][pos-1:pos-1+snv_len]
        if record.REF != ref_allele:
            return "ref_err"
            #raise Exception(\
            #     "ref allele in reference and in the VCF don't agree, they are "+ref_allele+" and "+record.REF+" respectively")
        # alleles = (ref,gt0,gt1)
        alleles = [ref_allele]
        for i in [0,2]:
            if gt is None:
                h = 'N'
            elif gt[i] == '0':
                h = ''.join(record.REF)
            elif gt[i] == '1':
                try:
                    h = "".join(record.ALT[0].sequence)
                except:
                    print record.ALT,type(record.ALT),record.ALT[0],type(record.ALT[0]),record.ALT[0].__dict__
                    raise
            elif int(gt[i]) >= 2:
                h = 'N'
                print chrom,record.POS, gt[i], "alleles, set to N"
            else:
                h = gt[i]
            alleles.append(h)
        return (alleles[0],alleles[1],alleles[2])
    
    #create info DataFrame
    vcf_region = vcf_to_validate.fetch(chrom,start,end)
    gen_id = [s for s in vcf_region.samples if ucla_id in s][0]
    #print vcf_region
    index = []
    allele0 = []
    allele1 = []
    ref_allele = []
    ref_err = []
    filters = []
    for record in vcf_region:
        pos = int(record.POS)
        alleles = get_alleles(record,gen_id)
        if alleles == "ref_err":
            ref_err.append(record.POS)
            continue
        if omit_non_segregating:
            if ((alleles[1] == alleles[0] and alleles[2] == alleles[0]) or (alleles[1] == 'N' and alleles[2] == 'N')):
                continue
        if omit_indels:
            if (len(alleles[0])>1 or len(alleles[1])>1 or len(alleles[2])>1):
                continue
        index.append(pos)
        ref_allele.append(alleles[0])
        allele0.append(alleles[1])
        allele1.append(alleles[2])
        filters.append(record.FILTER)
    #try:  
    #print allele1
    df = pd.DataFrame({'a0':allele0,'a1':allele1,'ref':ref_allele,'filter':filters},index=index)
    m_index = pd.MultiIndex.from_arrays([[chrom]*len(df.index),df.index.values])
    df.index = m_index
    return df, ref_err
```

```python
def snp_df(vcf,name='NN',max_interval=inf):
    df_tot = pd.DataFrame()
    ref_err = {chrom:[] for chrom in chromosomes}
    for interval in range(min(max_interval,len(pacbio_intervals_useable))):
        start =  pacbio_intervals_useable.iloc[interval]['start']
        end =  pacbio_intervals_useable.iloc[interval]['end']
        chrom = pacbio_intervals_useable.iloc[interval]['chrom']
        df, ref_err0 = get_snp_df(vcf,chrom,start,end,ucla_id)
        ref_err[chrom]+=ref_err0
        df_tot = df_tot.append(df)
    m_index = pd.MultiIndex.from_arrays([[name]*len(df_tot.columns),df_tot.columns.values])
    df_tot.columns = m_index
    for chrom in ref_err.keys():
        if ref_err[chrom]:
            print name,chrom, ref_err[chrom]
    return df_tot.reindex(pd.MultiIndex.from_tuples(df_tot.index.values[:]))
```

```python
def remove_non_segregating(df):
    """
    takes as input a pandas DataFrame with the same columns
    for multiple sequencing or genotyping methods
    remove sites that are not segregating in any of the methods 
    (SNPs that are equal to ref in the individual considered)
    """
    #names of the sequencin methods (0-level of df columns)
    seq_meths = set(df.columns.get_level_values(0).values)
    #list of boolean dfs giving segregating sites in each sequencing method
    seg = [df[s]['ref'].notnull()*((df[s]['a0']!=df[s]['ref'])+(df[s]['a1']!=df[s]['ref'])) for s in seq_meths]
    return df[reduce(lambda x,y: x+y,seg)]
    #df[df[]]
    #for sm in seq_meths:
        
```

```python
def remove_filtered_SNPs(df):
    """
    remove snps that are filtered or absent in all sequencing methods
    """
    seq_meths = set(df.columns.get_level_values(0).values)
    non_filt = [df[s]['filter'].apply(lambda x: not x) for s in seq_meths]
    return df[reduce(lambda x,y: x+y,non_filt)]
```

```python
def remove_all_filtered_SNPs(df):
    """
    remove filtered or set them to NaN even if the SNP is found in another method
    """
    def filtered_to_nan(series,col):
        if not not series[col]['filter']:
            series[col]=np.nan
        else:
            pass
        return series

    seq_meths = set(df.columns.get_level_values(0).values)
    for s in seq_meths:
        #print df[s]
        #print df[s][df[s]['filter'].notnull()*df[s]['filter'].apply(lambda x: not not x)].head()
        df = df.apply(lambda x: filtered_to_nan(x,s),axis=1)
    
    return df
        
```
