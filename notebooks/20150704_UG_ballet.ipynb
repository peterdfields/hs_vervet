```python
%%px --local
#general functions
def get_gen_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)

def get_subpop_df(gen_df,pop,poptype='pop3'):
    gen_df_pop = gen_df[mt[(mt[poptype]==pop)].index]
    #print len(gen_df_pop)
    #removing non-seggretating sites
    gen_df_pop = gen_df_pop[(gen_df_pop.mean(axis=1)>0)&(gen_df_pop.mean(axis=1)<2)]
    #print len(gen_df_pop)
    return gen_df_pop

def get_id_order(chrom):
    fn = jn(var_ana_dir,"163_UG_ref3500_ref_alt_pass_{}.012".format(chrom))
    with open(fn) as f:
        ids = f.readline().strip().split('\t')[2:]
    return ids


def get_ad_df(chrom):
    """
    get 0,1,2 genotype data frame
    """
    fn = jn(var_ana_dir,"163_UG_ref3500_anc_der_including_fixed_{}.tsv".format(chrom))
    df = pd.read_csv(fn,index_col=[0,1],sep="\t",na_values="N")
    return df.astype(np.float16)

def get_ad_seg_df(chrom):
    df = get_ad_df(chrom)
    return df[(df.sum(axis=1)>0)&\
                        (df.sum(axis=1)<2*df.notnull().sum(axis=1))]
```
