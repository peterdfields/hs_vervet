### Run ms

```python
seq_length = 10000
p = subprocess.Popen("ms 2 1 -t 1000 -r 500 " + str(seq_length) + " -T",shell=True,stdout=subprocess.PIPE,stderr=subprocess.PIPE)
o, e = p.communicate()
trees = o[o.index("//")+3:o.index("segsites")-1].split("\n")
segment_lengths = [int(t.split("]")[0][1:]) for t in trees]
tmrcas = [float(t.split(":")[1].split(",")[0]) for t in trees]
#test whether sequence length is consistent with sum of the segment lengthes
assert reduce(lambda x,y:x+y,segment_lengths) == int(seq_length)
tmrca_sequence = [[tmrca]*sl for tmrca,sl in zip(tmrcas,segment_lengths)]
tmrca_sequence = reduce(lambda x,y:x+y,tmrca_sequence)
snp_pos_rel = map(float,o[o.index("positions")+len("positions")+1:].split("\n")[0].split())
snp_pos_abs = (seq_length * np.array(snp_pos_rel)).astype(int)
```

```python
def hist(ibs):
    """
    returns tuple (hist,bin_edges)
    """
    imax = int(ceil(log(max(ibs))/log(1.25)))
    imin= int(floor(log(max(min(ibs),1))/log(1.25)))
    bins = [1.25**i for i in range(imin,imax)]
    return np.histogram(ibs,bins=bins)
```
